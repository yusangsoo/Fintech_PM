# 반복문
종류 <br>
for, while

* 반복문은 동일한 작업을 여러 번 반복해야 할 경우

* for, while 두 종류의 반복문이 있음

* for는 반복 횟수가 정해져 있는 경우에 사용

* while은 반복 횟수가 정해지지 않은 경우에 사용

for 반복문 : 반복할 횟수가 정해져 있는 경우

for 변수 in 반복할 객체, 자료:
____반복할 코드

순서가 있는 자료형은 모두 for문제 사용 가능

* 문자열

* 리스트

* 튜플

```python
a = "파이썬은 문자열을 반복문에 넣어 출력 가능"
for char in a:
  print(char, end = " ")
```

```python
for item in [1,2,3,4,5]:
  print(item, end = " ")
```

```python
L = [1,2,3,3, '문자열', [1,2,3], (4,5,6), {'반복문':'for'}]
for lis in L:
  print(lis, end = " ")
```

```python
d = dict(번호 = 1, 주소 = '서울', 전화번호 = '010-2222-1111', 이름 = '홍길동')
print(d)
```

```python
# 딕셔너리는 반복문에 넣었을 때 key만 출력된다

for item in d : 
  print(item, end = " ")
```

```python
for key in d :
  print(d[key], end = " ")
```

```python
for key in d :
  print(d.get(key,0), end= " ")
```

```python
j = 0
for i in range(1,11) :
  j = j + i
print(j)
```

```python
# 1부터 100까지 곱한 수를 출력
m = 1
for num in range(1,100) :
  m = m*num
print(m)
```

```python
# 1부터 50까지 짝수인 수만 더한 값
even = 0
for i in range(1,51) :
  if i % 2 == 0 :
    even = even + i
print(even)
```

```python
# 1부터 50까지 짝수인 수만 더한 값
even = 0
for i in range(1,26) :
    even = even + i * 2
print(even)
```

```python
# 1부터 100까지의 수 중에서 7의 배수만 찾아 곱한 값을 출력
seven = 1
for i in range(7,101,7) :
  print(i, end = " ")
  seven = seven*i
print(seven)
```

```python
# 1부터 100까지 수를 만들고 그중 30에서 70까지만 더해서 출력
f = 0
for i in range(1,101) :
  if i < 30 or i > 70 :
    i = 0
  f = f + i
print(f)
```

```python
# 1부터 100까지 수를 만들고 그중 30에서 70까지만 더해서 출력
temp = 0
for num in range(1,101)[29:70]:
  temp += num
print(temp)
```

```python
L1 = list(range(1,10,2))
print('L1', L1)
L2 = list(range(0,10,2))
print('L2', L2)

for num, num2 in zip(L1, L2):
  print(num, num2)
  temp = num+num2
  print(temp)
```

```python
# enumerate(): index를 만들어주는 함수
for index, char in enumerate("파이썬은 문자열을 반복문에 넣어 출력 가능") :
  print(index, char)
```

```python
# zip함수는 튜플로 결과를 묶어서 내놓음.
## 거기에 enumerate함수로 덮으니 zip함수의 결과를 튜플 1세트로 보는 것
### 그래서 zip함수에 몇 개가 있든 결국 튜플 1세트
for num1, num2 in enumerate(zip(L1,L2)) :
  print(num1, num2, num2[0])
```

```python
L3 = [5,4,2]
for index, (num1, num2, num3) in enumerate(zip(L1,L2,L3)) :
  print(index, num1)
```

```python
for num in reversed(range(1,11)):
  print(num)
```

### while
* 반복할 횟수가 정해져 있지 않은 경우
* 무한 반복이 가능해 그렇게 되지 않도록 조건식을 짜는 것이 중요 변수 선언 및 초기화 <br>
while 조건식 : <br>
__반복할 코드 <br>
__변수 증가/감소 로직 <br>

```python
# while문으로 1-10까지 출력
i = 1
while i <= 10 :
  print(i, end = " ")
  i += 1
```

```python
# while문을 이용해서 1~10까지 더해서 결과 출력
j = 0
i = 1
while i <= 10 :
  j = j+i
  i += 1
print(j)
```

```python
# 20에서 1까지 숫자를 더한 값을 출력
i = 20
hap = 0
while i > 0 :
  hap += i
  i -= 1
print(hap)
```

```python
# 1에서 50까지 수 중에서 3의 배수이고, 짝수인 수를 찾아서 더한 값 출력
i = 1
j = 0
while i <= 50 :
  if i % 3 == 0 and i % 2 == 0 :
    j += i
  i += 1
print(j)
```

```python
# 무한반복을 주는 법은 while에 true라고 조건을 준다.
# 무한반복을 주는 이유는 횟수를 모르기 때문
# 무한반복을 탈출시켜주는 코드(break)를 반드시 입력해야 한다.
```

```python
i = 1
hap = 0
while True:
  if i <= 100 :
    hap += i
    i += 1
  else:
    break
print(hap)
```

* True가 되는 경우 : 정수, 실수, 문자열, True를 넣을 때
* False가 되는 경우 : 0, 0.0, '', "", [], (), {}

### 반복문에서 아무일도 하지 않고 건너뛰기 : pass
* 우선 반복할 코드가 들어가야 하는 곳을 만들고 나중에 작업하기 위해 사용

```python
for i in range(10) : 
    pass
```

```python
# i += 1을 넣지 않으면 무한반복됨
i = 1
while i < 10 :
    pass
    i += 1
```

## break : 반복문을 정지시키는 명령

```python
for i in range(1,101) : 
    if i > 50 :
        break
    else:
        print(i, end = " ")
```

```python
i = 1
while i <= 100 :
    if i > 50 : 
        break
    else :
        print(i, end = " ")
    i+= 1
```

## continue : 특정 조건에서 코드 실행 건너뛰기

```python
# 1~100까지 수 중에 짝수면 더하고 홀수면 건너뛰기
num = 0
for i in range(1,101) : 
    if i % 2 == 0 :
        num += i
    else:
        continue
print(num)
    
```

#### 1~100까지 수 중에서 짝수면 더하고 홀수면 건너뛰는 while 반복문

```python
i = 1
num = 0
while i <= 100 :
    if i % 2 == 0 :
        num += i
        i += 1
    else:
        i+= 1
        continue
print(num)
```

```python
# 더 간소화
i = 1
num = 0
while i <= 100 :
    if i % 2 == 0 :
        num += i
    i+= 1
#     continue
print(num)
```

## 중첩반복문
* 두 개 이상의 반복문을 겹쳐서 사용
* 안쪽 반복이 끝난 후 바깥쪽 반복문이 실행

```python
for i in range(0,24) : 
    for j in range(0, 60) :
        print(i,"시", j, '분')
```

```python
# 매 반복마다 0.1초 쉬어라
import time
for i in range(0,24) : 
    for j in range(0, 60) :
        time.sleep(0.1)
        print(i,"시", j, '분', end = "\r")
```

```python
# 구구단
for first in range(1,10) : 
    for second in range(1,10) :
        print(first, "*", second, "=", first*second)
```

* 리스트에 동물 이름이 저장되어 있다.
* 리스트 = ['dog', 'cat', 'parrot']
* for 문을 이용해서 동물 이름의 첫 글자만 출력

```python
animal = ['dog', 'cat', 'parrot']
for i in animal :
    print(i[0])
```

* 리스트에서 세 글자 이상의 문자를 화면에 출력

```python
char = ['I','study', 'python','language','!']
for i in char :
    if len(i) >= 3 :
        print(i)
```

* 리스트에 저장된 데이터를 아래와 같이 출력
* apart = [[101. 102], [201,202], [301,302]]
* 결과 -> 101 호, 102 호, 201 호, 202 호, 301 호, 302 호

```python
apart = [[101, 102], [201,202], [301,302]]
for i in apart :
    for j in i :
        print(j, "호")
```

```python
apart = [[101, 102], [201,202], [301,302]]
for i in reversed(apart) :
    for j in i :
        print(j, "호")
```

```python
for i in range(len(apart)) :
    for j in range(len(apart)-1) :
        print(apart[i][j])
```
