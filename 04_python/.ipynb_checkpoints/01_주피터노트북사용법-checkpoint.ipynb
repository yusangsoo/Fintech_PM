# 파이썬 기초 문법
* 파이썬은 세미콜론을 사용하지 않는다
* 한 줄에 여러 명령을 쓸 때 사용
* 파이썬의 주석은 문장 앞에 # 을 붙인다
* 파이썬은 들여쓰기 사용
* 들여쓰기는 4칸이 기본(스페이스바 사용), google colab은 2칸
* 변수를 만들 때 스네이크 표기법 사용 -> user_name
* 클래스를 만들 땐 카멜 표기법 사용 -> UserName

# print() 문

```python
print('Hello, World')
```

```python
## 구분자
print('hello, World', 'hello python', sep='--')
```

```python
## 프린트가 끝난 후 어떤 작업을 할지
## defalut는 '\ㅜ' = 줄바꿈
print('hello, world', end = '')
print('hello, python')
```

```python
## 줄을 수행하고 왼쪽으로 다시 가서 작성
## 크롤링을 할 때 어디까지 진행됐는지 보기 위해 계속 덮어쓰며 진행상황 체크할 때 유용
print('hello, world', end = '\r')
print('hello, python')
```

# 변수
* 영문, 숫자를 조합해서 만들 수 있다.
* 문자부터 시작해야 하며 숫자부터 시작하면 안됨
* 대소문자 구별
* 언더바 _ 로 시작 가능
* 특수문자 사용 안됨
* 예약어(if, for while, str 등)는 사용할 수 없다.
* 변수는 스네이크 표기법

# 키보드 입력받기input()
* input()함수로 입력받는 모든 값은 문자열로 저장
* 반드시 형변환이 필요

```python
num1 = input()
num2 = input()
print(int(num1)+int(num2))
```

```python
num1, num2 = input().split()
print(int(num1)+int(num2))
```
