# 분석에 사용할 기술적 지표 만들기
- MA_45 : 45일 단순 이동평균 : 특정 자산의 45일간 평균 종가
- VMA_45 : 45일 거래량 : 지난 45일 간의 평균 거래량
- RSI_14 : 14일 상대강도지수(Relative Strength Index) : 14일 동안의 자산 가격 변동을 바탕으로 과매수 또는 과매도 상태를 평가하는 지표

## RSI 공식
- RSI는 자산의 가격 변동 강도를 측정해 과매수, 또는 과매도 상태를 평가하는 기술적 분석 지표
- 주로 14일 동안의 가격 변동 기준으로 계산. 값은 0 - 100 사이
- RSI 값이 70 이상 : 과매수 상태(매도 시점을 고려)
- RSI 값이 30 이하 : 과매도 상태(매수 시점을 고려)
- RSI는 주가가 너무 빠르게 상승하거나 하락했는지 확인하는데 사용
- 투자자들의 매수/매도 결정을 내릴 때 중요한 참고 지표로 활용
### RSI 계산 공식
$$ RSI = {100 - {100 \over 1 + RS}} $$
RS(Relative Strength)
$$ RS = {Average Gain \over Average Loss} $$
- Average Gain : 일정 기간(14일)동안 가격 상승분의 평균
- Average Loss : 일정 기간(14일)동안 가격 하락분의 평균
- 상승 Gain = 오늘의 종가 - 어제의 종가
- 하락 Loss = 어제의 종가 - 오늘의 종가

```python
days = 45
```

단순이동평균 구해서 데이터 프레임에 합치기

```python
# 종가 45일 평균
ma = pd.Series(data['CLOSE_SPY'].rolling(days).mean(), name = 'MA_' + str(days))
ma
```

```python
data = data.join(ma)
data
```

```python
# 거래량 이동평균 vma
vma = pd.Series(data['VOLUME'].rolling(days).mean(), name = 'VMA_' + str(days))
vma
```

```python
data = data.join(vma)
data
```

```python
# RSI 지수 구하기
## diff를 통해 변화량 구하기
delta = data['CLOSE_SPY'].diff()
delta
```

```python
# 상승
gain = (delta.where(delta > 0, 0)).rolling(window = 14).mean()
# 하락
loss = (-delta.where(delta < 0, 0)).rolling(window = 14).mean()
```

```python
# RS 구하기
RS = gain / loss
RS
```

```python
# RSI 구하기
## RSI 공식에 따라 RSI 값 계산
### 70 이상이면 과매수, 30이하면 과매도
RSI = 100 - (100 / (1 + RS))
RSI
```

```python
RSI.name = 'RSI_14'
```

```python
data = data.join(RSI)
data
```

```python
data = data.dropna()
```

타겟 변수 생성 pct_change 변동률을 구해주는 함수

```python
data.loc[:, 'pct_change'] = data['CLOSE_SPY'].pct_change()
data
```

```python
# 수익이 났으면 1, 손해가 나면 0으로 변경
data.loc[:, 'target'] = np.where(data['pct_change'] > 0, 1, 0)
data
```

```python
data = data.dropna()
data
```

- 미래 예측을 위해 타겟 변수를 shift(-1)을 주기
- 현재 행의 주가 데이터를 다음 행의 주가 데이터로 미리 이동시켜 오늘 데이터를 기반으로 내일 주가를 예측하는 구조

```python
# 미래 예측을 위해 타겟 변수를 shift(-1)을 주기
data['target'] = data['target'].shift(-1)
data
```

```python
data = data.dropna()
data
```

```python
data.columns
```

```python
# 가격을 예측하는 것이 아니라 오를건지 내릴건지만 보는 것
X = data.drop(['CLOSE_SPY', 'OPEN', 'HIGH', 'LOW', 'VOLUME', 'pct_change', 'target'], axis = 1)
y = data['target']
```

```python
X
```

## 시계열 데이터의 홀드아웃
- 시계열 데이터이기에 홀드아웃시 날짜가 섞이면 안됨
- train_test_split 옵션에서 shuffle 옵션을 반드시 False로 해야 함

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, shuffle = False)
```

```python
X_train
```

```python
y_train
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
```

```python
xgb = XGBClassifier(n_estimators = 400, learning_rate = 0.1, max_depth = 3, random_state = 7)
xgb.fit(X_train, y_train)
pred = xgb.predict(X_test)
print(classification_report(y_test, pred))
```
