```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python
data = pd.read_csv('./data/creaditcard_fraud_data.csv')
data
```

```python
# 데이터 타입 확인
data.info()
```

```python
data = data.drop(['first', 'last', 'street', 'city', 'state', 'zip', 'trans_num', 'unix_time'], axis = 1)
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data, stratify = data['is_fraud'], test_size = 0.4, random_state = 7)
train
```

# 카드 번호에 따른 평균 지출액, 표준편차 구하기

```python
train.groupby('cc_num')['amt'].agg(['mean', 'std']).reset_index()
```

```python
amt_by_cc_num =  train.groupby('cc_num')['amt'].agg(['mean', 'std']).reset_index()
amt_by_cc_num
```

```python
train = pd.merge(train, amt_by_cc_num, how = 'left', on = 'cc_num')
train
```

# 이상지출 탐지하기
- 평소 사용자가 지출하던 금액보다 과도하게 높은 경우
- z 스코어를 이용해 탐지
- z 스코어 : 값 - 평균 / 표준편차

```python
train['amt_z_score'] = (train['amt'] - train['mean']) / train['std']
```

```python
train[['amt', 'mean', 'std', 'amt_z_score']]
```

```python
train[train['amt_z_score'] > 3]
```

```python
train['age'] = train['dob'].apply(lambda x : 2020-int(x[:4]))
train.head()
```

# 위도 경도 정보로 거리 계산하기
- 카드 사용자의 주거지 위치와 카드 사용처의 거리를 계산해 너무 멀면 사기일 가능성이 높다

```python
# !pip install geopy
```

```python
import geopy.distance
```

```python
train['customer_coord'] = pd.Series(zip(train['lat'], train['long']))
train['merchant_coord'] = pd.Series(zip(train['merch_lat'], train['merch_long']))
```

```python
train['distance'] = train.apply(lambda x : geopy.distance.distance(x['customer_coord'], x['merchant_coord']).km, axis = 1)
train.head()
```

```python
spend_by_distance = train.groupby('cc_num')['distance'].agg(['mean', 'std']).reset_index()
spend_by_distance
```

```python
train = pd.merge(train, spend_by_distance, how = 'left', on = 'cc_num')
train
```

```python
train['distance_z_score'] = (train['distance'] - train['mean']) / train['std']
train.head()
```

```python
train = train.drop(['mean', 'std'], axis = 1)
train
```

```python
train = train[['gender', 'age', 'amt', 'amt_z_score', 'category_z_score', 'is_fraud_rate', 'distance_z_score', 'is_fraud']]
```

```python
train = train.dropna()
```

```python
X = train.drop('is_fraud', axis = 1)
y = train['is_fraud']
```

```python
X = pd.get_dummies(X, drop_first = True)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, stratify = y, test_size = 0.4, random_state = 7)
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
scaled_X = rs.fit_transform(X_train)
scaled_X = pd.DataFrame(scaled_X, columns = X_train.columns)
scaled_X_valid = rs.fit_transform(X_valid)
scaled_X_valid = pd.DataFrame(scaled_X_valid, columns = X_train.columns)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score
```

```python
xgb = XGBClassifier(max_depth = 5, n_estimators = 500, n_jobs = 10, random_state = 7)
xgb.fit(scaled_X, y_train)
valid_pred = xgb.predict(scaled_X_valid)
print(classification_report(y_valid, valid_pred))
print(roc_auc_score(y_valid, valid_pred))
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
```

```python
pipe = Pipeline([
    ('rs', RobustScaler()),
    ('smt', SMOTE(random_state = 10)),
    ('rfc', RandomForestClassifier(n_jobs = 10, random_state = 10))
])
```

```python
# 랜덤 서치를 위한 파라미터 정의
params_rv = {
    'rfc__n_estimators' : [100, 500, 1000],
    'rfc__max_depth' : [3, 5, 7]
}
```

```python
# 교차 검증을 위한 stratify-kfold
scv = StratifiedKFold(n_splits = 5, random_state = 7, shuffle = True)
```

```python
random_cv = RandomizedSearchCV(pipe, param_distributions = params_rv, n_iter = 10,
                              scoring = 'roc_auc', cv = scv, random_state = 10, n_jobs = 10)
random_cv.fit(X_train, y_train)
print('best parameter : ', random_cv.best_params_)
print('best score : ', random_cv.best_score_)
print('best model test score : ', random_cv.score(X_valid, y_valid))
```

```python
models = dict(XGB = XGBClassifier(n_jobs = 5, random_state = 7),
             RFC = RandomForestClassifier(n_jobs = 5, random_state = 7),
             LGBM = LGBMClassifier(n_jobs = 5, random_state = 7))
models
```

```python
params = {
    'XGB' : {
        'XGB__max_depth' : [3, 5, 7],
        'XGB__n_estimators' : [100, 300, 500],
        'XGB__subsample' : [0.6, 0.8 ,1.0]
    },
    'RFC' : {
        'RFC__max_depth' : [3, 5, 7],
        'RFC__n_estimators' : [100, 300, 500]
    },
    'LGBM' : {
        'LGBM__max_depth' : [3, 5, 7],
        'LGBM__n_estimators' : [100, 300, 500],
        'LGBM__subsample' : [0.6, 0.8 ,1.0]
    }
}
```

```python
# 각 모델별로 최적의 파라미터 찾고 성능 평가
best_estimators = {}
roc_auc_scores = {}
classfication_reports = {}
```

```python
skf = StratifiedKFold(n_splits = 5, random_state = 10, shuffle = True)
```

```python
for model_name, model in models.items() :
    print(model_name, model)
    pipe = Pipeline([('rs', RobustScaler()),
                    ('smt', SMOTE(random_state=10))
                     ,(model_name, model)])
    rand_cv = RandomizedSearchCV(pipe, param_distributions = params[model_name], cv = skf, scoring = 'roc_auc', n_jobs = 5)
    rand_cv.fit(X_train, y_train)
    print('best parameters : ', rand_cv.best_params_)
    best_model = rand_cv.best_estimator_
    best_pred = best_model.predict(X_valid)
    print('report : ', classification_report(y_valid, best_pred))
    print('rocauc : ', roc_auc_score(y_valid, best_pred))
```
