```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python
data = pd.read_csv('./data/tr_eod_data_rounded.csv', index_col = 0, parse_dates = True)
data
```

- 로그수익률 데이터 만들기

```python
rets = np.log(data / data.shift(1)) # 로그수익률을 만드는 공식
rets
```

```python
rets.cumsum().apply(np.exp).plot(figsize = (12, 6))
plt.show()
```

- 월간 수익률 그래프

```python
rets.cumsum().apply(np.exp).resample('1m', label = 'right').last().plot(figsize = (12, 6))
```

- 애플 주식에 대한 평균 수익률 구하고 그래프 그리기

```python
sym = 'AAPL.O'
data = pd.DataFrame(data[sym]).dropna()
data
```

- 20일 간의 애플주식 rolling 지표 구하기

```python
window = 20
```

```python
data['min'] = data[sym].rolling(window).min() # 20일 기간의 이동 최저값
data['mean'] = data[sym].rolling(window).mean() # 20일 기간의 이동 평균
data['std'] = data[sym].rolling(window).std() # 20일 기간의 이동 표준편차
data['median'] = data[sym].rolling(window).median() # 20일 기간의 이동 중앙값
data['mean'] = data[sym].rolling(window).mean() # 20일 기간의 이동 평균
data['max'] = data[sym].rolling(window).max() # 20일 기간의 이동 최대값
data['ewma'] = data[sym].ewm(halflife = 0.5, min_periods = window).mean() # 20일 기간의 이동 가중 평균
data
```

- 애플 주가 데이터 시각화 최근데이터 200개만 시각화

```python
ax = data[['min', 'mean', 'max']].iloc[-200:].plot(figsize = (12, 6), style = ['g--', 'r--', 'g--'], lw = 0.8)
ax
data[sym].iloc[-200:].plot(ax = ax, lw = 2.0)
plt.title('AAPL 20일 Moving Average Price Movement')
plt.show()
```

# 자본시장 시계열의 기술적 분석
- 단기 이동 평균을 42일로 - 2달
- 장기 이동 평균 252일로 - 1년

```python
data['short_ma'] = data[sym].rolling(window = 42).mean() # 단기 이동 평균
data['long_ma'] = data[sym].rolling(window = 252).mean() # 장기 이동 평균
```

```python
data[[sym, 'short_ma', 'long_ma']].plot(figsize = (12, 6))
plt.title('APPL 단순 이동 평균 2달 & 1년')
plt.legend()
plt.show()
```

## 골든 크로스(golden cross)
- 단기 이동평균선이 장기 이동 평균선을 아래에서 위로 교차할 때
- 상승장의 시작을 알리는 신호
    - 단기 이동 평균선 : 보통 50일 이동 평균선을 사용
    - 장기 이동 평균선 : 보통 200일 이동 평균선을 사용
## 데드 크로스(dead cross)
- 단기 이동 평균선이 장기 이동 펴균선을 위에서 아래로 교차할 때
- 하락장을 알리는 신호

```python
# 골든 크로스와 데드 크로스 그래프 같이 그리기
data['positions'] = np.where(data['short_ma'] > data['long_ma'], 1, -1) # 1일 땐 상승장, 매수. -1일 땐 하락장, 매도
ax = data[[sym, 'short_ma', 'long_ma', 'positions']].plot(figsize = (12, 6), secondary_y = 'positions')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
plt.title('AAPL Trading Window based on Technical Analysis')
plt.show()
```
