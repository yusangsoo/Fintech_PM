```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
import pmdarima as pm
from pmdarima.arima import ndiffs
```

```python
data = pd.read_csv('./data/tr_eod_data_rounded.csv', index_col = 0, parse_dates = True)
data
```

```python
data_y = data.dropna()
```

```python
y = data_y[['AAPL.O']].iloc[-200:]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
y_train, y_test = train_test_split(y['AAPL.O'], test_size = 0.3, shuffle = False)
```

```python
y_train
```

```python
y_train.plot()
y_test.plot()
```

# 정상성 검정 및 차분 수 구하기

```python
# alpha는 유의수준, test는 ['kpss' or 'adf'], max_d는 차분 횟수. 최대가 6
kpss_diffs = ndiffs(y_train, alpha = 0.05, test = 'kpss', max_d = 6)
adf_diffs = ndiffs(y_train, alpha = 0.05, test = 'adf', max_d = 6)
print(kpss_diffs)
print(adf_diffs)
```

```python
# 정상 시계열로 만들기 위한 차분 횟수
n_diffs = max(kpss_diffs, adf_diffs)
n_diffs
```

```python
plt.plot(y_train.diff(1))
```

```python
model = pm.auto_arima(y_train, trace = True)
```

```python
model.summary()
```

prob(Q)가 0.05보다 클 때 정상성을 말함

```python
model.plot_diagnostics(figsize = (12, 6))
plt.show()
```

```python
# y_test 기간 = 60일
y_pred = model.predict(n_periods = len(y_test))
```

```python
y_pred
```

- 결과가 모두 같은 값으로 나온 이유 : 하루 예측하고 예측한 값도 다시 train에 넣어 모델을 업데이트 시켜야 함

## 모델 업데이트

```python
def forecast_one_step() :
    fc, conf_int = model.predict(n_periods = 1, return_conf_int = True)
    print(fc, conf_int)
    return (fc.tolist()[0], np.asarray(conf_int).tolist()[0])
```

```python
forecasts = []
y_pred = []
pred_upper = []
pred_lower = []

for new_ob in y_test :
    fc, conf = forecast_one_step()
    y_pred.append(fc)
    pred_upper.append(conf[1])
    pred_lower.append(conf[0])
    ## 모델 업데이트
    model.update(new_ob)
print(y_pred)
```

```python
final_result = pd.DataFrame(dict(test = y_test, pred = y_pred))
final_result
```

```python
fig, axes = plt.subplots(1,1, figsize = (12, 6))
plt.plot(y_train, label = 'Train')
plt.plot(y_test, label = 'Test')
plt.plot(final_result['pred'], label = 'Prediction')
plt.legend()
plt.show()
```
