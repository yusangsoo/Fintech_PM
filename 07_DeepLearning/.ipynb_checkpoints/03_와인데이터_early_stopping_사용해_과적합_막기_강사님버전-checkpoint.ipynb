```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine = pd.read_csv("./data/winequality-white.csv", sep=";")
wine
```

```python
wine.info()
```

```python
wine.describe()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
X = wine.drop('quality', axis = 1).copy()
y = wine['quality'].copy()
y2 = wine['quality'].copy()
```

```python
rs = RobustScaler()
X_scaled = rs.fit_transform(X)
```

```python
y = pd.get_dummies(y)
y
```

```python
# !pip install imblearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_scaled, y, test_size = 0.4, stratify = y, random_state = 7)
X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size = 0.5, stratify = y_valid, random_state = 7)
```

```python
print(X_train.shape, X_valid.shape, X_test.shape)
```

데이터 증폭

```python
from imblearn.over_sampling import ADASYN
```

```python
adasyn = ADASYN(random_state = 10, n_neighbors = 2, n_jobs = 10)
X_train_adasyn, y_train_adasyn = adasyn.fit_resample(X_train, y_train)
```

```python
X_train_adasyn = pd.DataFrame(X_train_adasyn, columns = X.columns)
X_train_adasyn
```

```python
y_train_adasyn = pd.get_dummies(y_train_adasyn)
y_valid = pd.get_dummies(y_valid)
y_test = pd.get_dummies(y_test)
y_train_adasyn
```

```python
from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
model = Sequential()
model.add(Dense(32, input_dim = X_train.shape[1], activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(7, activation = 'softmax'))
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
early_stop = EarlyStopping(patience = 100)
filepath = './model/white_wine{epoch:04d}__{val_loss:.4f}.keras'
model_save = ModelCheckpoint(filepath = filepath, save_best_only = True)
history = model.fit(X_train_adasyn, y_train_adasyn, epochs = 10000, batch_size = 500, validation_data = (X_valid, y_valid),
                   callbacks = [early_stop, model_save])
```

```python
wine_best_model = load_model('./model/white_wine0027__1.2883.keras')
wine_pred = wine_best_model.predict(X_test)
wine_pred = pd.DataFrame(wine_pred, columns = y_test.columns)
```

```python
wine_pred_class = wine_pred.idxmax(axis = 1)
wine_pred_class
```

```python
y_test_class = y_test.idxmax(axis = 1)
y_test_class = y_test_class.reset_index(drop = True)
y_test_class
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test_class, wine_pred_class))
```

XGB로 비교분석

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y2_labeled = le.fit_transform(y2)
y2_labeled
```

```python
type(y2_labeled)
```

```python
X_train2, X_valid2, y_train2, y_valid2 = train_test_split(X_scaled, y2_labeled, test_size = 0.4, stratify = y2, random_state = 7)
X_valid2, X_test2, y_valid2, y_test2 = train_test_split(X_valid2, y_valid2, test_size = 0.4, stratify = y_valid2, random_state = 7)
```

```python
adasyn = ADASYN(random_state = 10, n_neighbors = 2, n_jobs = 10)
X_train2_adasyn, y_train2_adasyn = adasyn.fit_resample(X_train2, y_train2)
```

```python
# !pip install xgboost
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier(max_depth = 5, n_estimators = 1000, n_jobs = 10, random_state = 7)
xgb.fit(X_train2_adasyn, y_train2_adasyn)
xgb_pred = xgb.predict(X_valid2)
print(classification_report(le.inverse_transform(y_valid2), le.inverse_transform(xgb_pred)))
```
