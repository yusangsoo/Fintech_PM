## 실제 사용할 때 함수화해 사용

```python
def data_preprocessing(data) :
    # 결측치 처리하는 것도 여기서 처리해야 함
    data = pd.get_dummies(data)
    data = data[['OverallQual', 'GrLivArea', 'GarageCars', 'YearBuilt', 'FullBath',
       'TotalBsmtSF', '1stFlrSF', 'ExterQual_Gd', 'YearRemodAdd', 'Fireplaces',
       'KitchenQual_TA', 'GarageFinish_Unf', 'ExterQual_TA']]
    rs = joblib.load('./model/iowa_robustscaler.joblib')
    data_scaled = rs.transform(data)
    data_scaled = pd.DataFrame(data_scaled, columns = data.columns)
    return data_scaled
```

```python
def model_use(data) : 
    model = load_model('./model/iowa_house439__1591623296.0000.keras')
    pred = model.predict(data)
    pred = pd.DataFrame(pred, columns = ['test_pred'])
    return pred
```

```python
data2 = pd.read_csv('./data/house_test.csv')
result = data_preprocessing(data2)
result = model_use(result)
result
```
