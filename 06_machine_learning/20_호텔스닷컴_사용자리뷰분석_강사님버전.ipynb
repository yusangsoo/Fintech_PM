## Naive Bayes를 이용한 감성분석

```python
def tokenizer(x) :
    token = mecab.morphs(x)
    return token
```

```python
X = data['description']
y = data['isgood']
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

### CountVectorizer

```python
cv = CountVectorizer(tokenizer = tokenizer, max_df = 0.8, min_df = 4, ngram_range = (1,3), stop_words = stopwords)
X_cv = cv.fit_transform(X)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_cv, y, test_size = 0.4, random_state = 7)
X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size = 0.5, random_state = 7)
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
pred = mnb.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
pred = mnb.predict(X_test)
print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

### TfidfVectorizer

```python
tv = TfidfVectorizer(tokenizer = tokenizer, max_df = 0.8, min_df = 4, ngram_range = (1,3), stop_words = stopwords)
X_tv = tv.fit_transform(X)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_tv, y, test_size = 0.4, random_state = 7)
X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size = 0.5, random_state = 7)
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
pred = mnb.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
pred = mnb.predict(X_test)
print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```
