```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/snkn1959/data_source/main/customer_pca.csv')
data
```

```python
X = data.drop('label', axis = 1)
y = data['label']
```

PCA를 이용해 11개의 변수를 2개로 축소

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)
```

```python
X_pca = pd.DataFrame(X_pca, columns = ['comp1', 'comp2'])
```

```python
X_pca
```

```python
customer_pca = pd.concat([X_pca, y], axis = 1)
customer_pca
```

```python
sns.scatterplot(x = 'comp1', y = 'comp2', data = customer_pca, hue = 'label', palette = 'rainbow')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, stratify = y, random_state = 8)
X_train_PCA, X_test_PCA, y_train_PCA, y_test_PCA = train_test_split(X_pca, y, test_size = 0.4, stratify = y, random_state = 8)
```

```python
X_train_PCA.head()
```

```python
for i in range(1, 10) :
    dtc1 = DecisionTreeClassifier(max_depth = i, random_state = 8)
    dtc1.fit(X_train, y_train)
    pred = dtc1.predict(X_test)
    print(i, classification_report(y_test, pred))
```

```python
dtc2 = DecisionTreeClassifier(max_depth = 3, random_state = 8)
dtc2.fit(X_train_PCA, y_train_PCA)
pred = dtc2.predict(X_test_PCA)
print(classification_report(y_test_PCA, pred))
```

```python
result = pd.DataFrame(dtc1.feature_importances_, index = dtc1.feature_names_in_)
result.sort_values(by = 0, ascending = False)
```

컬럼 많은 걸 PCA로 축소해 분석하기

```python
an = pd.read_csv('https://media.githubusercontent.com/media/musthave-ML10/data_source/main/anonymous.csv')
an
```

```python
an['class'].value_counts()
```

```python
X = an.drop('class', axis = 1)
y = an['class']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
X_scaled = ss.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns = X.columns)
X_scaled.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.4, random_state = 8)
```

```python
dtc = DecisionTreeClassifier(max_depth = 5, random_state = 8)
dtc.fit(X_train, y_train)
pred = dtc.predict(X_test)
print(classification_report(y_test, pred))
```

```python
result = pd.DataFrame(dtc.feature_importances_, index = dtc.feature_names_in_)
result.sort_values(by = 0, ascending = False)
```

```python
result[result[0] > 0]
```

```python
len(result[result[0] > 0])
```

PCA로 차원축소

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 20)
X_pca = pca.fit_transform(X_scaled)
X_pca = pd.DataFrame(X_pca)
X_pca
```

```python
print(pca.explained_variance_)
```

```python
print(pca.explained_variance_ratio_)
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size = 0.4, random_state = 8)
```

```python
dtc = DecisionTreeClassifier(max_depth = 5, random_state = 8)
dtc.fit(X_train_pca, y_train_pca)
pred = dtc.predict(X_test_pca)
print(classification_report(y_test_pca, pred))
```

```python
result2 = pd.DataFrame(dtc.feature_importances_)
result2.sort_values(by = 0, ascending = False)
```

scree plot 그리기

```python
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
X_pca = pd.DataFrame(X_pca)
```

```python
# 주성분의 분산 비율
explained_variance_ratio = pca.explained_variance_ratio_

# Scree plot 그리기
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.grid()
plt.show()

# 누적 분산 비율 계산
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# 최적의 주성분 개수 찾기 (예: 80% 이상의 분산을 설명하는 주성분 개수)
optimal_components = np.argmax(cumulative_variance_ratio >= 0.80) + 1
print(f'Optimal number of components: {optimal_components}')
```
