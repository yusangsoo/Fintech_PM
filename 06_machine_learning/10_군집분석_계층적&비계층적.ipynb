# 계층적 군집 분석
- 비지도학습
- 데이터가 적을 경우 사용
- 거리 계산 지표 : 유클리드 거리, 맨하탄, 코사인 유사도(범주형)
    - 단일 연결법 : 최단연결법이라고도 함. 군집과 군집 사이에서 가장 가까운 데이터를 기준으로 묶음
    - 완전 연결법 : 최장연결법이라고도 함. 군집과 군집 사이에서 가장 먼 데이터를 기준으로 묶음
    - 평균 연결법 : 평균연결법 : 군집과 군집 사이의 모든 데이터의 평균
    - 중심 연결법 : 군집의 중심점 사이의 거리를 거리로 측정한 방법, 계산이 빠르다.
    - 와드 연결법 : 군집 내 오차가 최소가 되는 데이터로 계산. 군집 내 분산이 최소
- 계층적 군집의 시각화는 덴드로그램으로 한다. (덴드로그램은 sklearn에 없음)

# 군집의 최적 개수 정하기
- elbow method
- 실루엣 지수

- elbow method : inertia_(응집도)를 출력하고 시각화한 뒤 응집도의 감소폭이 급격한 변화가 사라지는 지점을 최적군집의 수로 채택하는 것

```python
print(km.inertia_)
```

```python
distance = []
for i in range(1, 21) :
    km = KMeans(n_clusters = i)
    km.fit(data)
    distance.append(km.inertia_)
distance
```

```python
plt.figure(figsize = (20, 10))
sns.lineplot(x = range(1, 21), y = distance, marker = 'o', markersize = 10, markerfacecolor = 'red')
plt.show()
```

실루엣 지수를 출력해서 최적 군집 개수 구하기
- 실루엣 지수를 구하기 위해서는 최소 2개 이상의 군집이 필요

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_scores = []
for i in range(2, 21) :
    km2 = KMeans(n_clusters = i)
    km2.fit(data)
    labels = km2.labels_
    silhouette_scores.append(silhouette_score(data, labels))
silhouette_scores
```

```python
plt.figure(figsize = (20, 10))
sns.lineplot(x = range(2, 21), y = silhouette_scores, marker = 'o', markersize = 10, markerfacecolor = 'red')
plt.show()
```

# K-means로 군집분석을 할 때 주의점
- k-means는 평균값을 이용하고 주로 유클리드 거리를 사용하기 때문에 이상치에 민감.
- 군집분석을 하기 전에 데이터의 스케일을 미리 맞춰주는 것이 중요

```python
data3 = data.drop('cluster', axis = 1)
data3
```

```python
sns.boxplot(data3['합산요금'])
```

```python
X = data3.drop('이탈여부_Yes', axis = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
scaled_X = mm.fit_transform(X)
scaled_X
```

```python
scaled_X = pd.DataFrame(scaled_X, columns = X.columns)
scaled_X
```

```python
km4 = KMeans(n_clusters = 2)
km4.fit(scaled_X)
result4 = km4.labels_
result4
```

```python
scaled_X['cluster'] = result4
```

```python
print(accuracy_score(data3['이탈여부_Yes'], scaled_X['cluster']))
```

```python
distance = []
for i in range(1, 21) :
    km = KMeans(n_clusters = i)
    km.fit(scaled_X)
    distance.append(km.inertia_)
distance
```

```python
plt.figure(figsize = (20, 10))
sns.lineplot(x = range(1, 21), y = distance, marker = 'o', markersize = 10, markerfacecolor = 'red')
plt.show()
```

### DBSCAN : 밀도 기반 군집 분석 알고리즘

```python
from sklearn.cluster import DBSCAN
```

```python
final_df = scaled_X.copy()
```

```python
scaled_X = scaled_X.drop('cluster', axis = 1)
scaled_X
```

```python
model = DBSCAN(eps = 2, min_samples = 5)
DB_result = model.fit_predict(scaled_X)
DB_result
```

```python
final_df['db_cluster'] = DB_result
```
