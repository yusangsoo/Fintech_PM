## 학원데이터 우편번호로 우편번호 데이터에서 주소 추출

```python
new_pc_academy.columns
```

```python
new_pc_academy.info()
```

```python
 new_pc_academy['시설도로명우편번호'] = new_pc_academy['시설도로명우편번호'].astype(int)
```

```python
new_pc_academy = pd.merge(new_pc_academy, address_df, how = 'left', left_on = '시설도로명우편번호', right_on = '우편번호')
new_pc_academy
```

```python
old_pc_academy['시설도로명우편번호'] = old_pc_academy['시설도로명우편번호'].astype(int)
```

```python
old_pc_academy = pd.merge(old_pc_academy, old_address_df, how = 'left', left_on = '시설도로명우편번호', right_on = '우편번호')
old_pc_academy
```

```python
merged_academy = pd.concat([new_pc_academy, old_pc_academy])
merged_academy
```

```python
merged_academy = merged_academy.drop(merged_academy[merged_academy['우편번호'].isna() == True].index)
merged_academy
```

```python
merged_academy = merged_academy.reset_index(drop = True)
```

```python
merged_academy['시도'] = merged_academy['시도'].str.replace('강원도', '강원특별자치도').str.replace('전라북도', '전북특별자치도')
merged_academy['시도'].unique()
```

## 시군구 법정동명 별 학원 수
