```python
data = data.drop(['Email', 'Address', 'Avatar'], axis = 1)
data.head()
```

```python
data.columns = data.columns.str.replace(' ', '_')
```

```python
data.columns = data.columns.str.replace('.', '')
```

```python
data.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('Yearly_Amount_Spent', axis = 1)
```

```python
y = data['Yearly_Amount_Spent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

```python
# statsmodels R 문법을 사용하는 거
from statsmodels.formula.api import ols
```

```python
train_data = pd.concat([X_train, y_train], axis = 1)
train_data.head()
```

```python
lm = ols('Yearly_Amount_Spent ~ Avg_Session_Length + Time_on_App + Time_on_Website + Length_of_Membership + state_name',
         data = train_data).fit()
```

```python
lm.summary()
```

- 회귀 계수는 0.984
- p-value는 4.36e-279임. 0.05보다 작음
- Avg_Session_Length(25.4119), Time_on_App(61.7833), Length_of_Membership(61.7833)의 p-vlaue가 0.05보다 작음
- AR(1.2649), MI(9.6009), MT(12.1776)도 0.05보다 작음

y = Avg_Session_Length * 25.4119 + Time_on_App * 61.7833 + Length_of_Membership * 61.7833 + state_name[T.AP] * 1.2649 + state_name[T.MI] * 9.6009 + state_name[T.MT] * 12.1776 - 1044.0261

```python
pred = lm.predict(X_test)
pred
```

```python
# 테스트가 얼마나 맞았나 보기
from sklearn.metrics import mean_squared_error
```

```python
print('Mean_Squared_Error(MSE) : ', mean_squared_error(y_test,pred))
print('Roo_Mean_Squared_Error(RMSE) : ', np.sqrt(mean_squared_error(y_test,pred)))
```

```python
# p-value 가져오기
p_value = lm.pvalues[1]
```

```python
# 산점도와 회귀선 그리기
plt.figure(figsize=(10,6))
sns.scatterplot(x = y_test, y = pred, label = 'data_points')
sns.regplot(x = y_test, y = pred, scatter = False, label = "Regresstion Line", color = 'red')
plt.xlabel('Actual Yearly Amount Spent')
plt.ylabel('pred Yearly Amount Spent')
plt.legend()
plt.show()
```

```python
X = pd.get_dummies(X, drop_first = True)
X
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
X_scaled = mms.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns = X.columns)
```

```python
X_scaled.head()
```

```python
# statsmodels의 파이썬 문법
import statsmodels.api as sm
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_scaled, y, test_size = 0.2, random_state = 7)
```

```python
# 상수항 추가
X_train2 = sm.add_constant(X_train2)
X_test2 = sm.add_constant(X_test2)
```

```python
X_train2.head(2)
```

```python
X_train2 = pd.get_dummies(X_train2)
X_train2
```

```python
X_test2 = pd.get_dummies(X_test2)
X_test2
```

```python
lr = sm.OLS(y_train2, X_train2).fit()
```

```python
lr.summary()
```

```python
pred2 = lr.predict(X_test2)
```

```python
print('MSE : ', mean_squared_error(y_test2, pred2))
print('RMSE : ', np.sqrt(mean_squared_error(y_test2, pred2)))
```

```python
p_value2 = lr.pvalues[1]
```

```python
# 산점도와 회귀선 그리기
plt.figure(figsize=(10,6))
sns.scatterplot(x = y_test2, y = pred2, label = 'data_points')
sns.regplot(x = y_test2, y = pred2, scatter = False, label = "Regresstion Line", color = 'red')
plt.xlabel('Actual Yearly Amount Spent')
plt.ylabel('pred Yearly Amount Spent')
plt.legend()
plt.show()
```
