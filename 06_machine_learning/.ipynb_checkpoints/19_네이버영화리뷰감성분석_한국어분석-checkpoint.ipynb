```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
```

```python
train_df = pd.read_csv("https://raw.githubusercontent.com/haram4th/ablearn/main/ratings_train.txt", sep='\t')
test_df = pd.read_csv("https://raw.githubusercontent.com/haram4th/ablearn/main/ratings_test.txt", sep='\t' )
```

```python
string.punctuation
```

```python
ss = string.punctuation + 'ㅡ' + '♥' + 'ㅋ' + 'ㅠ' + 'ㅜ' + 'ㅈ'
ss
```

```python
def remove_punc(x) :
    cleaned = []
    for i in x :
        if i not in ss :
            cleaned.append(i)
    cleaned = ''.join(cleaned)
    return cleaned
```

```python
train_df = train_df.dropna()
test_df = test_df.dropna()
```

```python
train_df['document'] = train_df['document'].apply(remove_punc)
```

```python
test_df['document'] = test_df['document'].apply(remove_punc)
```

```python
X_train_df = train_df['document']
X_test_df = test_df['document']
```

```python
y_train_df = train_df['label']
y_test_df = test_df['label']
```

```python
whole_words = []
for i in X_train_df :
    for word in i.split() :
        whole_words.append(word)
```

```python
whole_words
```

벡터화 해보기

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
X_train_vec = cv.fit_transform(X_train_df)
X_train_vec
```

```python
cv = CountVectorizer()
X_test_vec = cv.fit_transform(X_test_df)
X_test_vec
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import MultinomialNB
```

```python
X_train, X_train_test, y_train, y_train_test = train_test_split(X_train_vec, y_train_df, test_size = 0.2, random_state = 7)
# X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size = 0.5, random_state = 7)
```

```python
mnb = MultinomialNB()
mnb.fit(X_train_vec, y_train_df)
pred = mnb.predict(X_train_test)
print(accuracy_score(y_train_test, pred))
print(classification_report(y_train_test, pred))
```

```python
pred = mnb.predict(X_test_vec)
print(accuracy_score(y_test_df, pred))
print(classification_report(y_test_df, pred))
```
