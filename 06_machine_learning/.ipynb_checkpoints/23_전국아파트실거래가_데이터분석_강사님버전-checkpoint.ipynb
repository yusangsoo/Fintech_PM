# 전국 주소 정보 - 우체국 전국 우편번호 데이터

```python
# glob은 경로상에 있는 모든 파일 경로를 리스트로 만들어 줌
from glob import glob
```

```python
address_path = glob('../과제/model/5_zipcode/*.txt')
address_path
```

```python
address_df = pd.DataFrame()
for area in address_path : 
    temp = pd.read_csv(area, sep = '|')
    address_df = pd.concat([address_df, temp])
address_df
```

```python
address_df.info()
```

```python
address_df = address_df[['우편번호', '시도', '시군구', '법정동명', '읍면', '리명']]
address_df
```

```python
address_df = address_df.drop_duplicates(subset = ['우편번호'])
address_df = address_df.reset_index(drop = True)
address_df.head()
```

```python
address_df = address_df.fillna(' ')
address_df.info()
```

# 전국 학원 정보 불러오기

```python
academy = pd.read_csv('../과제/model/academy.csv', encoding = 'cp949')
academy.head()
```

```python
academy.info()
```

```python
academy['시설도로명우편번호'] = academy['시설도로명우편번호'].apply(lambda x : str(x).strip())
```

### 우편번호 패턴 파악

```python
# 신우편번호 범위
print(address_df['우편번호'].min())
print(address_df['우편번호'].max())
```

```python
# 구우편번호 범위
print(old_address_df['우편번호'].min())
print(old_address_df['우편번호'].max())
```

```python
new_pc_academy = academy[(academy['시설도로명우편번호'].str.len() == 4) | (academy['시설도로명우편번호'].str.len() == 5)]
new_pc_academy
```

```python
old_pc_academy = academy[academy['시설도로명우편번호'].str.len() == 6]
old_pc_academy
```

## 학원데이터 우편번호로 우편번호 데이터에서 주소 추출

```python
new_pc_academy.columns
```

```python
new_pc_academy.info()
```

```python
 new_pc_academy['시설도로명우편번호'] = new_pc_academy['시설도로명우편번호'].astype(int)
```

```python
new_pc_academy = pd.merge(new_pc_academy, address_df, how = 'left', left_on = '시설도로명우편번호', right_on = '우편번호')
new_pc_academy
```

```python
old_pc_academy['시설도로명우편번호'] = old_pc_academy['시설도로명우편번호'].astype(int)
```

```python
old_pc_academy = pd.merge(old_pc_academy, old_address_df, how = 'left', left_on = '시설도로명우편번호', right_on = '우편번호')
old_pc_academy
```

```python
merged_academy = pd.concat([new_pc_academy, old_pc_academy])
merged_academy
```

```python
merged_academy = merged_academy.drop(merged_academy[merged_academy['우편번호'].isna() == True].index)
merged_academy
```

## 시군구 읍명동 별 학원 수

```python
academy_num = merged_academy.groupby(['시도', '시군구', '읍면동', '리'])[['학원지정번호']].count()
academy_num
```

```python
academy_num = academy_num.reset_index()
academy_num = academy_num.rename(columns = {'학원지정번호' : '학원수'})
academy_num
```
