```python
data = data.drop(['education', 'capital-gain', 'capital-loss'], axis = 1)
data.head()
```

```python
data = pd.get_dummies(data,
               columns = ['age', 'workclass', 'marital-status', 'occupation', 'relationship', 'race','sex', 'native-country',],
               drop_first = True)
data.head()
```

```python
data['class'] = data['class'].apply(lambda x : 1 if x == " >50K" else 0)
```

```python
data.head()
```

```python
data['over50byage'] = data['over50byage'].astype('int')
```

```python
data.info()
```

```python
X = data.drop('class', axis = 1)
y = data['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.4, stratify = y, random_state = 7)
X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size = 0.5, stratify = y_valid, random_state = 7)
```

```python
X_valid
```

```python
28791 + 9597
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
dtc = DecisionTreeClassifier(max_depth = 3, random_state = 7)
dtc.fit(X_train, y_train)
pred = dtc.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
for i in range(1, 10) : 
    dtc = DecisionTreeClassifier(max_depth = i, random_state = 7)
    dtc.fit(X_train, y_train)
    pred = dtc.predict(X_valid)
    print(i, accuracy_score(y_valid, pred))
    print(i, classification_report(y_valid, pred))
```

```python
dtc = DecisionTreeClassifier(max_depth = 6, random_state = 7)
dtc.fit(X_train, y_train)
pred = dtc.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
# 실제 테스트하기 때문에 fit은 하지 않음
pred = dtc.predict(X_test)
print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

## 배깅으로 돌리기
- 부트스트랩
- 투표

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bcf = BaggingClassifier(n_estimators = 1000, n_jobs = 10, random_state = 7)
bcf.fit(X_train, y_train)
pred = bcf.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
best = DecisionTreeClassifier(max_depth = 6, random_state = 7)
```

```python
bcf = BaggingClassifier(estimator = best, n_estimators = 1000, n_jobs = 10, random_state = 7)
bcf.fit(X_train, y_train)
pred = bcf.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

## 부스팅 모델
- 에이다 부스트
- 그래디언트 부스트
- xg 부스트

```python
from xgboost import XGBClassifier
```

```python
xg = XGBClassifier(n_estimators = 1000, max_depth = 3, random_state = 7, n_jobs = 10, learning_rate = 0.1)
xg.fit(X_train, y_train)
pred = xg.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
pred = xg.predict(X_test)
print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

```python
from lightgbm import LGBMClassifier
```

```python
lgb = LGBMClassifier(n_estimators = 1000, max_depth = 3, random_state = 7, n_jobs = 10, learning_rate = 0.1, verbose = -1)
lgb.fit(X_train, y_train)
pred = lgb.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
pred = lgb.predict(X_test)
print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```
