```python
X = data.drop('주택가격', axis = 1)
y = data['주택가격']
```

데이터 스케일링
- standard : 표준 정규분포 - 데이터의 특성이 사라짐. 이상치에 취약
- minmax : 모든 데이터를 0과 1사이의 숫자로 스케일링. 데이터의 특성을 유지하지만 이상치에 취약
- robust : 사분위수를 이용해서 스케일링. 이상치에 민감하지 않음


```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
X_scaled = rs.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns = X.columns)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state = 7)
```

```python
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# 상수항 추가
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)
```

```python
ols = sm.OLS(y_train, X_train).fit()
pred = ols.predict(X_test)
print('mean_absolute_error(MAE) : ', mean_absolute_error(y_test, pred))
print('mean_squared_error(MSE) : ', mean_squared_error(y_test, pred))
print('root_mean_squared_error(RMSE) : ', np.sqrt(mean_squared_error(y_test, pred)))
print('r2_score(r-squared) : ', r2_score(y_test, pred))
```

```python
print(ols.summary())
```

# 1차 회귀 분석 후 유의하지 않은 컬럼 제거 후 재분석

```python
X_scaled2 = X_scaled.drop('비소매상업지역비율', axis = 1)
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_scaled2, y, test_size=0.2, random_state = 7)
```

```python
# 상수항 추가
X_train2 = sm.add_constant(X_train2)
X_test2 = sm.add_constant(X_test2)
```

```python
ols2 = sm.OLS(y_train2, X_train2).fit()
pred2 = ols2.predict(X_test2)
print('mean_absolute_error(MAE) : ', mean_absolute_error(y_test2, pred2))
print('mean_squared_error(MSE) : ', mean_squared_error(y_test2, pred2))
print('root_mean_squared_error(RMSE) : ', np.sqrt(mean_squared_error(y_test2, pred2)))
print('r2_score(r-squared) : ', r2_score(y_test2, pred2))
```

```python
print(ols2.summary())
```

```python
plt.figure(figsize=(20, 10))
sns.regplot(x = y_test2, y = pred2, scatter = True)
plt.show()
```

## sklearn에 있는 linear로 한 번 돌려보기

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
lr_result = pd.DataFrame(lr.coef_, index = lr.feature_names_in_)
display(lr_result)
print("intercept : ", lr.intercept_)
print('mean_absolute_error(MAE) : ', mean_absolute_error(y_test, pred))
print('mean_squared_error(MSE) : ', mean_squared_error(y_test, pred))
print('root_mean_squared_error(RMSE) : ', np.sqrt(mean_squared_error(y_test, pred)))
print('r2_score(r-squared) : ', r2_score(y_test, pred))
```

## 라쏘(L1 제약, 절댓값으로 제약, 변수를 분석에서 제외)
- 모델의 과적합을 방지하고 범용성을 높이는 것에 목적을 둠

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso()
lasso.fit(X_train, y_train)
pred = lasso.predict(X_test)
lasso_result = pd.DataFrame(lasso.coef_, index = lasso.feature_names_in_)
display(lasso_result)
print("intercept : ", lasso.intercept_)
print('mean_absolute_error(MAE) : ', mean_absolute_error(y_test, pred))
print('mean_squared_error(MSE) : ', mean_squared_error(y_test, pred))
print('root_mean_squared_error(RMSE) : ', np.sqrt(mean_squared_error(y_test, pred)))
print('r2_score(r-squared) : ', r2_score(y_test, pred))
```
