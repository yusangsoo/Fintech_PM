```python
import string
```

```python
string.punctuation
```

```python
def remove_punc(x) :
    cleaned = []
    for i in x :
        if i not in string.punctuation :
            cleaned.append(i)
    cleaned = ''.join(cleaned)
    return cleaned.replace("  ", " ").replace('\n', '').replace("   ", " ")
```

```python
data['text'] = data['text'].apply(remove_punc)
```

# 상투적인 단어 제거하기
- nltk 라는 라이브러리로 상투적인 단어(stopwords) 제거하기

```python
# !pip install nltk
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
print(stopwords.words('english'))
```

```python
stopwords2 = stopwords.words('english')
```

```python
stopwords2.extend(['They', 'they', 'non', 'also', 'I', 'The', 'place', 'We', 'would', 'get', 'This'])
```

```python
def remove_stopwords(x) :
    cleaned = []
    for i in x.split() :
        if i not in stopwords2 :
            cleaned.append(i)
    cleaned = ' '.join(cleaned)
    return cleaned
```

```python
X = X.apply(remove_stopwords)
```

# 자주 등장하는 단어 빈도 분석

```python
whole_words = []
for i in X :
    for word in i.split() :
        whole_words.append(word)
```

```python
len(whole_words)
```

```python
from nltk.probability import FreqDist
```

```python
# 빈도수 확인
FreqDist(whole_words)
```

```python
# 빈도수 그래프 그리기(50개만)
plt.figure(figsize = (20, 10))
FreqDist(whole_words).plot(50)
```

# 나이브베이즈를 이용해 감성분석 하기
- 텍스트를 벡터화해야 함

## text를 벡터화하기 countVectorizer
- 단어의 빈도를 세고 숫자로 변형

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
X_vec = cv.fit_transform(X)
X_vec
```

```python
type(X_vec)
```

```python
for i in X_vec[:1] :
    print(i)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_vec, y, test_size = 0.4, random_state = 7)
X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size = 0.5, random_state = 7)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
pred = mnb.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
pred_t = mnb.predict(X_test)
print(accuracy_score(y_test, pred_t))
print(classification_report(y_test, pred_t))
```

### randomforest로 돌려보기

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 500, max_depth = 5, n_jobs = 10, random_state = 7)
rfc.fit(X_train, y_train)
pred2 = rfc.predict(X_valid)
print(accuracy_score(y_valid, pred2))
print(classification_report(y_valid, pred2))
```
