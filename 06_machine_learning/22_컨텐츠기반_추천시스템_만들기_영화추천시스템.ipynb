```python
import json
```

```python
# name 값을 추출하는 함수
def extract_names(genres_str):
    if genres_str == '[]':
        return []
    genres = json.loads(genres_str)  # JSON 파싱
    return [genre['name'] for genre in genres]  # name 값 추출
```

```python
data['genres'].apply(lambda x: extract_names(x))
```

```python
# genres 컬럼에 대치
data['genres'] = data['genres'].apply(lambda x: extract_names(x))
```

```python
data
```

```python
# genres 컬럼에 대치
data['keywords'] = data['keywords'].apply(lambda x: extract_names(x))
data['production_companies'] = data['production_companies'].apply(lambda x: extract_names(x))
data['production_countries'] = data['production_countries'].apply(lambda x: extract_names(x))
data['spoken_languages'] = data['spoken_languages'].apply(lambda x: extract_names(x))
```

```python
data['genres'] = data['genres'].apply(lambda x : ' '.join(x))
data['keywords'] = data['keywords'].apply(lambda x : ' '.join(x))
data['production_companies'] = data['production_companies'].apply(lambda x : ' '.join(x))
data['production_countries'] = data['production_countries'].apply(lambda x : ' '.join(x))
data['spoken_languages'] = data['spoken_languages'].apply(lambda x : ' '.join(x))
```

- genres의 단어를 벡터화
- 코사인유사도를 구해 비슷한 장르의 행 검색
- 유사도가 높은 행을 기준으로 내림차순 정렬

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range = (1, 2))
genres_mat = cv.fit_transform(data['genres'])
print(genres_mat)
```

```python
genres_mat.shape
```

코사인 유사도를 사용해 genres_mat의 유사도를 산출 후 genres_sim이라는 컬럼 생성
- 비슷한 장르의 영화를 찾기 위해 코사인 유사도를 계산
- 코사인 유사도는 주로 문자, 문서의 유사도를 계산하는데 사용됨

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
genres_sim = cosine_similarity(genres_mat, genres_mat)
print(genres_sim.shape)
```

```python
print(genres_sim[:10])
```

```python
sorted_genres_sim = genres_sim.argsort()[:,::-1]
```
