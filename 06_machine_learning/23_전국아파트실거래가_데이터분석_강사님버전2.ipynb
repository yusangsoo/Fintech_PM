```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python
apart_df = pd.read_csv('./data/아파트실거래데이터.csv')
apart_df
```

### apart_df를 train과 test로 나누어 저장하기

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(apart_df, test_size = 0.4, random_state = 8)
```

### train데이터 전처리

```python
train.head()
```

```python
train.isna().sum()
```

```python
train = train.dropna()
```

```python
train = train.reset_index(drop = True)
```

```python
X = train.drop(['거래금액', 'date'], axis = 1)
y = train['거래금액']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le2 = LabelEncoder()
X['주소'] = le.fit_transform(X['주소'])
X['단지명'] = le2.fit_transform(X['단지명'])
X.head()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
X_scaled = rs.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns = X.columns)
X_scaled
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_scaled, y, test_size = 0.4, random_state = 7)
X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size = 0.5, random_state = 7)
```

### xgboost 돌려보기

```python
from xgboost import XGBRegressor
```

```python
xgbr = XGBRegressor(max_depth = 5, n_estimators = 1000, n_jobs = 10, random_state = 7)
xgbr.fit(X_train, y_train)
xgbr_pred = xgbr.predict(X_valid)
print('mean_squared_error : ', mean_squared_error(y_valid, xgbr_pred))
print('rmse : ', np.sqrt(mean_squared_error(y_valid, xgbr_pred)))
print('r2-score : ', r2_score(y_valid, xgbr_pred))
```

```python
xgbr_test_pred = xgbr.predict(X_test)
print('mean_squared_error : ', mean_squared_error(y_test, xgbr_test_pred))
print('rmse : ', np.sqrt(mean_squared_error(y_test, xgbr_test_pred)))
print('r2-score : ', r2_score(y_test, xgbr_test_pred))
```

```python
plt.figure(figsize = (10, 20))
plt.plot(y_test[200000:200100].values, label = '실제값')
plt.plot(xgbr_test_pred[200000:200100], label = '예측값')
plt.xlabel('samples')
plt.ylabel('values')
plt.title('실제값과 예측값')
plt.legend()
plt.show()
```
