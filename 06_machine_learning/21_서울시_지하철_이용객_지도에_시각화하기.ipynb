```python
# 서울시 지하철 이용객 정보 파일
data = pd.read_csv('./data/seoul-metro-2021.logs.csv')
data.head()
```

```python
# 서울 지하철 역 정보 파일
station_info = pd.read_csv('./data/seoul-metro-station-info.csv')
station_info.head()
```

```python
station_info = station_info[['station.code', 'line.name', 'station.name_full', 'geo.latitude', 'geo.longitude']]
station_info
```

# data와 station_info의 컬럼 이름 동일하게 처리하기
- 컬럼 이름 변경하기
- station_code를 인덱스로 바꾸기

```python
station_info.columns = ['station_code', 'line_name', 'station_name_full', 'geo_latitude', 'geo_longitude']
station_info.columns
```

```python
data = data.set_index('station_code')
data
```

```python
station_info = station_info.set_index('station_code')
station_info
```

```python
total = station_info.join(data)
total
```

# 각 역 별 승하차 인원 구하기

```python
people_in_out_by_station = total.groupby(['station_code', 'station_name_full', 'geo_latitude', 'geo_longitude'])[['people_in', 'people_out']].sum()
people_in_out_by_station
```

```python
people_in_out_by_station = people_in_out_by_station.reset_index().set_index('station_code')
people_in_out_by_station
```

```python
morning.head()
```

```python
morning_sum = morning.groupby('station_code')[['people_in', 'people_out']].sum()
evening_sum = evening.groupby('station_code')[['people_in', 'people_out']].sum()
```

```python
morning_sum
```

```python
morning_sum = morning_sum.join(station_info[['geo_latitude', 'geo_longitude']])
evening_sum = evening_sum.join(station_info[['geo_latitude', 'geo_longitude']])
```

```python
seoul_morning_in = folium.Map(location = [37.55, 126.98], zoom_start = 12)
HeatMap(data = morning_sum[['geo_latitude', 'geo_longitude', 'people_in']]).add_to(seoul_morning_in)
seoul_morning_in
```

```python
seoul_morning_out = folium.Map(location = [37.55, 126.98], zoom_start = 12)
HeatMap(data = morning_sum[['geo_latitude', 'geo_longitude', 'people_out']]).add_to(seoul_morning_out)
seoul_morning_out
```

```python
seoul_evening_in = folium.Map(location = [37.55, 126.98], zoom_start = 12)
HeatMap(data = evening_sum[['geo_latitude', 'geo_longitude', 'people_in']]).add_to(seoul_evening_in)
seoul_evening_in
```

```python
seoul_evening_out = folium.Map(location = [37.55, 126.98], zoom_start = 12)
HeatMap(data = evening_sum[['geo_latitude', 'geo_longitude', 'people_out']]).add_to(seoul_evening_out)
seoul_evening_out
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
scaled = mms.fit_transform(total[['people_in', 'people_out']])
scaled = pd.DataFrame(scaled, columns = ['people_in', 'people_out'])
scaled
```

```python
total = total.reset_index()
total
```

```python
total['people_in'] = scaled['people_in']
total['people_out'] = scaled['people_out']
total
```

```python
from folium.plugins import HeatMapWithTime
```

```python
heat_data = []
for timestamp, group in total.groupby('timestamp') :
    heat_map_data = []
    for _, row in group.iterrows() :
        heat_map_data.append([row['geo_latitude'], row['geo_longitude'], row['people_in']])
    heat_data.append(heat_map_data)
```

```python
heat_data[:3]
```

```python
m2 = folium.Map(location = [37.56, 126.97], zoom_start = 10)
HeatMapWithTime(heat_data, radius = 15).add_to(m2)
m2
```

```python
index = total.groupby('timestamp')[['people_in', 'people_out']].sum().index
index = index.strftime('%Y-%m-%d %H:%M:%S').tolist()
index
```

```python
m2 = folium.Map(location = [37.56, 126.97], zoom_start = 10)
HeatMapWithTime(heat_data, index = index, radius = 15, control=True).add_to(m2)
m2
```
