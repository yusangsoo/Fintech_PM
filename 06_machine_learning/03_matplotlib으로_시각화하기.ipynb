```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

## 간단한 그래프 예제

```python
# 그래프를 그릴 도화지를 준비
fig, ax = plt.subplots()

# 앞 리스트는 x축, 뒤 리스트는 y축
ax.plot([1,2,3,4], [1,4,2,3])

# 그래프 보기
plt.show()
```

- figure : 전체 그래프가 그려지는 도화지
- axes : 한 칸에 그려지는 그래프
- axis : 그래프의 x,y축

```python
# 그래프가 없이 도화지만 그릴 수 있음
fig = plt.figure()
plt.show()
```

x, y축이 하나씩 있는 그래프 그리기

2 * 2로 4개 그래프 그리기

## matplotlib의 2가지 코딩 스타일

### oo-스타일 -matplotlib의 공식 추천 스타일
- 복잡한 그래프를 그릴 때 사용
- 사용법이 pyplot 스타일보다 복잡함

```python
x = np.linspace(0,2,100)
x
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
# 파란 선
ax.plot(x, x, label = 'linear')
# 주황 선
ax.plot(x, x**2, label = 'quadratic')
# 초록 선
ax.plot(x, x**3, label = 'cubic')
# x 축 라벨
ax.set_xlabel('x label')
# y 축 라벨
ax.set_ylabel('y label')
# 제목
ax.set_title('simple plot')
# 범주
ax.legend()
plt.show()
```

## pyplot 스타일
- 00-style 보다 사용법이 간단함
- 아주 복잡한 그래프를 그리기엔 부적합

```python
x = np.linspace(0,2,100)
```

```python
plt.figure(figsize=(5, 2.7))
plt.plot(x, x, label = 'linear')
plt.plot(x, x**2, label = 'quadratic')
plt.plot(x, x**3, label = 'cubic')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('simple plot pyplot style')
plt.legend()
plt.show()
```

### 선 모양 바꾸기
- set_linestyle(모양)

```python
# 정수로 숫자를 랜덤하게 만듦
data1, data2, data3, data4 = np.random.randn(4,100)
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))
# cumsum은 누적합
# linestyle은 '-', '--', '-.', ':', '' 이렇게 5개 가능
ax.plot(x, np.cumsum(data1), color = 'blue', linewidth = 3, linestyle = '--')
i, = ax.plot(x, np.cumsum(data2), color = 'orange', linewidth = 2)
i.set_linestyle = ':'
plt.show()
```

### 색 바꾸기
- facecolor
- edgecolor

```python
fig, ax = plt.subplots(figsize = (5, 2.7))
ax.scatter(data1, data2, s = 50, facecolor = 'm', edgecolor = 'k')
plt.show()
```

### 선 넓이, 선 형태, 마커 모양과 크기

```python
fig, ax = plt.subplots(figsize= (5, 2.7))
ax.plot(data1, 'o', label = 'data1')
ax.plot(data2, 'd', label = 'data2')
ax.plot(data3, 'v', label = 'data3')
ax.plot(data4, 'x', label = 'data4')
ax.legend()
plt.show()
```

### x축 & y축 이름, 눈금, 그래프 이름

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7))
n, bins, patches = ax.hist(x, 50, density = True, facecolor = "C0", alpha = 0.75)
# x축 제목 입력
ax.set_xlabel("length [cm]")
# y축 제목 입력
ax.set_ylabel("probability")
# 제목 입력
ax.set_title("Aardvark lengths \n (not really)")
# 텍스트 입력
ax.text(75, .025, r'$\mu=115, \ \sigma=15$')
# x, y 축의 끝 값 지정 후 보여주기
ax.axis([55, 175, 0, 0.03])
# 그리드 주기
ax.grid(True)
plt.show()
```

### 눈금자

```python
fig, ax = plt.subplots(1,2, figsize = (5, 2.7))
xdata = np.arange(len(data1))
data = 10**data1
ax[0].plot(xdata, data)
ax[1].plot(xdata, data)
ax[1].set_yscale('log')
plt.show()
```

눈금자 편집

```python
fig, ax = plt.subplots(2, 1, figsize = (5, 2.7), layout = 'constrained')
ax[0].plot(xdata, data1)
ax[0].set_title('Automatic ticks')

ax[1].plot(xdata, data1)
ax[1].set_xticks(np.arange(0,100,30), ['zero', '30', 'sixty', '90'])
ax[1].set_yticks([-1.5, 0, 1.5])
ax[1].set_title('manual ticks')
plt.show()
```

그래프 안에 그래프 겹쳐 그리기

```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)

fig, (ax1, ax3) = plt.subplots(1, 2, figsize = (5, 2.7), layout = 'constrained')
l1, = ax1.plot(t,s)
ax2 = ax1.twinx()
l2, = ax2.plot(t, range(len(t)), "C1")
ax2.legend([l1, l2], ['sine(left)', 'straight (right)'])

ax3.plot(t,s)
ax3.set_xlabel("Angle [rad]")
ax4 = ax3.secondary_xaxis('top', functions=(np.rad2deg, np.deg2rad))
ax4.set_xlabel("Angle [˚]")
plt.show()
```
