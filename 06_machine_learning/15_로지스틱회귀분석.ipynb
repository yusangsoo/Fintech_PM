# 로지스틱 회귀분석
- odds <br> $ odds = {p \over 1-p} $ 
<br><br>
- 로짓변환 <br> $logit = \log{p \over 1-p}$
<br><br>
- 시그모이드 함수로 변환되어 0, 1의 분류 분석 가능

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/haram4th/ablearn/main/Taitanic_train.csv")
data
```

```python
X = data[['Pclass', 'Sex', 'Age']]
y = data['Survived']
```

```python
X = pd.get_dummies(X, drop_first = True)
X
```

```python
X['Age'] = X['Age'].fillna(29.0)
```

```python
X
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
```

```python
lg = LogisticRegression()
lg.fit(X_train, y_train)
pred = lg.predict(X_test)
print("accuracy_score : ", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
print('roc_auc_score : ', roc_auc_score(y_test, pred))
```

```python
fper, tper, _ = roc_curve(y_test, pred)
plt.plot(fper, tper)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
X_train_scaled = mms.fit_transform(X_train)
X_test_scaled = mms.fit_transform(X_test)
```

```python
import statsmodels.api as sm
```

```python
lg2 = sm.GLM(y_train, X_train_scaled, family = sm.families.Binomial())
res = lg2.fit()
res.summary()
```

```python
# 상수항 추가
X_train_scaled = sm.add_constant(X_train_scaled)
X_test_scaled = sm.add_constant(X_test_scaled)
```

```python
model = sm.Logit(y_train, X_train_scaled).fit()
pred = model.predict(X_test_scaled)
print(model.summary())
```

```python
pred = np.round(pred)
```

```python
print(accuracy_score(y_test, pred))
```
