```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mem = pd.read_csv("http://raw.githubusercontent.com/haram4th/ablearn/main/member.csv")
tran = pd.read_csv("http://raw.githubusercontent.com/haram4th/ablearn/main/transaction.csv")
```

```python
mem.head()
```

- conversion : 쿠폰 사용여부
- recency = 최근방문일 = 현재일 - 최근 구매일
- zip_code = 사는 곳
- channel : 방문하는 주채널

```python
tran.head()
```

- num_item : 구매 갯수
- total_amount : 총 금액

```python
tran['sum_num_item'] = tran.groupby('id')['num_item'].transform(sum)
tran.head()
```

```python
tran.info()
```

```python
tran['sum_total_amount'] = tran.groupby('id')['total_amount'].transform(sum)
tran.head()
```

```python
tran['count_num_item'] = tran.groupby('id')['num_item'].transform('count')
tran.head()
```

```python
# 한 번 방문했을 때 얼마를 쓰는지
tran['mean_count_amount'] = tran['sum_total_amount'] / tran['count_num_item']
# 구매 물건의 평균 가격
tran['mean_sum_num_amount'] = tran['sum_total_amount'] / tran['sum_num_item']
tran.head()
```

```python
tran_dist = tran.drop(['num_item', 'total_amount'], axis = 1)
tran_dist.head()
```

```python
tran_dist = tran_dist.drop_duplicates(subset=['id'])
```

```python
tran_dist.head()
```

```python
mdf = pd.merge(mem, tran_dist, how = 'inner', on = 'id')
mdf.head()
```

```python
mdf.describe()
```

```python
mdf = pd.get_dummies(mdf)
```

```python
mdf.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = mdf.drop(['id', 'conversion'], axis = 1)
y = mdf['conversion']
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.4, stratify = y, random_state = 7)
X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size = 0.5, stratify = y_valid, random_state = 7)
```

```python
y_train.value_counts()
```

```python
y.value_counts()
```

## 의사결정나무

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
for i in range(1, 10) : 
    dtc = DecisionTreeClassifier(max_depth = i, random_state = 7)
    dtc.fit(X_train, y_train)
    pred = dtc.predict(X_valid)
    print(i, accuracy_score(y_valid, pred))
    print(i, classification_report(y_valid, pred))
```

```python
dtc = DecisionTreeClassifier(max_depth = 9, random_state = 7)
dtc.fit(X_train, y_train)
pred = dtc.predict(X_valid)
print(i, accuracy_score(y_valid, pred))
print(i, classification_report(y_valid, pred))
```

```python
pred = dtc.predict(X_test)
print(i, accuracy_score(y_test, pred))
print(i, classification_report(y_test, pred))
```

## 배깅으로 돌리기

```python
from sklearn.ensemble import BaggingClassifier
```

```python
best = DecisionTreeClassifier(max_depth = 9, random_state = 7)
```

```python
bcf = BaggingClassifier(n_estimators = 1000, n_jobs = 10, random_state = 7)
bcf.fit(X_train, y_train)
pred = bcf.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
bcf = BaggingClassifier(estimator = best, n_estimators = 1000, n_jobs = 10, random_state = 7)
bcf.fit(X_train, y_train)
pred = bcf.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
pred = bcf.predict(X_test)
print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

## 그리드서치로 해보기

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
xgb_params = dict(
    n_estimators = [100, 500, 1000, 1500],
    max_depth = [3,5,7],
    learning_rate = [0.1, 0.5, 1.0],
    n_jobs = [10],
    random_state = [7],
    sampling_method = ['hist', 'uniform', 'gradient_based']
)
```

```python
xgb = XGBClassifier()
```

```python
grid_cv = GridSearchCV(xgb, param_grid = xgb_params, cv = 5, n_jobs = 10, verbose = 3)
grid_cv.fit(X_train, y_train)
pred = grid_cv.predict(X_valid)
print(sorted(grid_cv.cv_results_.keys()))
print('최적 하이퍼파라미터 : ', grid_cv.best_params_)
print(f"최고 예측 정확도 : {grid_cv.best_score_:.4f}")
```

```python
xg = XGBClassifier(n_estimators = 100, max_depth = 5, random_state = 7, n_jobs = 10, learning_rate = 0.1, sampling_method = 'uniform')
xg.fit(X_train, y_train)
pred = xg.predict(X_valid)
print(i, accuracy_score(y_valid, pred))
print(i, classification_report(y_valid, pred))
```

기존 XGBoost보다 더 낮은 정확도를 나타냄

## 랜덤포레스트로 돌려보기

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc_params = dict(
    n_estimators = [100, 500, 1000, 1500],
    max_depth = [3,5,7],
    n_jobs = [10],
    random_state = [7],
    criterion = ['gini', 'entropy', 'log_loss']
)
```

```python
rfc_cv = RandomizedSearchCV(rfc, param_distributions = rfc_params, n_iter = 100, cv = 5, n_jobs = 10, verbose = 3, random_state = 7)
rfc_cv.fit(X_train, y_train)
pred = rfc_cv.predict(X_valid)
print(sorted(rfc_cv.cv_results_.keys()))
print('최적 하이퍼파라미터 : ', rfc_cv.best_params_)
print(f"최고 예측 정확도 : {rfc_cv.best_score_:.4f}")
```

```python
rfc2 = RandomForestClassifier(n_estimators = 100, max_depth = 7, n_jobs = 10, random_state = 7)
rfc2.fit(X_train, y_train)
pred2 = rfc2.predict(X_valid)
print(accuracy_score(y_valid, pred2))
print(classification_report(y_valid, pred2))
```

## 데이터 증폭하기

```python
smt = SMOTE(sampling_strategy = 'auto', random_state = 7, k_neighbors = 5, n_jobs = 10)
X_train_smt, y_train_smt = smt.fit_resample(X_train, y_train)
```

```python
rfc_cv = RandomizedSearchCV(rfc, param_distributions = rfc_params, n_iter = 100, cv = 5, n_jobs = 10, verbose = 3, random_state = 7)
rfc_cv.fit(X_train_smt, y_train_smt)
pred = rfc_cv.predict(X_valid)
print(sorted(rfc_cv.cv_results_.keys()))
print('최적 하이퍼파라미터 : ', rfc_cv.best_params_)
print(f"최고 예측 정확도 : {rfc_cv.best_score_:.4f}")
```
