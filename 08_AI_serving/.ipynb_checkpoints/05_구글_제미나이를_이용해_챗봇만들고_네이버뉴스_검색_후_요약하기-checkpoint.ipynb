```python
import gradio as gr
import os
import requests
import pandas as pd
from datetime import datetime

# 제미나이 모델 초기화 (모델 객체 생성)
model = genai.GenerativeModel("gemini-1.5-flash")

# 문자열에서 불필요한 태그 제거
def string_delete(text):
    return text.replace("<b>", "").replace("</b>", "").replace("&quot;", "")

# 날짜 문자열 파싱
def parse_date(date_str):
    return datetime.strptime(date_str, "%a, %d %b %Y %H:%M:%S %z")

# 데이터프레임에서 태그 제거 및 날짜 파싱
def make_df(df_name):
    for i in df_name:
        for idx in range(0, len(df_name[i])):
            df_name[i][idx] = string_delete(df_name[i][idx])
    df_name['pubDate'] = df_name['pubDate'].apply(parse_date)
    return df_name

# 네이버 뉴스 API에서 뉴스 목록 가져오기
def fetch_related_news(keyword):
    client_id = os.getenv('id')  # 환경 변수에서 네이버 클라이언트 ID 가져오기
    client_pw = os.getenv('pw')  # 환경 변수에서 네이버 클라이언트 Secret 가져오기
    
    url = "https://openapi.naver.com/v1/search/news.json"
    payload = {'query': keyword, 'display': 20, 'start': 1, 'sort': 'date'}  # 뉴스 검색 쿼리
    headers = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_pw}
    
    # API 요청
    r = requests.get(url, params=payload, headers=headers)
    if r.status_code == 200:
        data = r.json()
        df = pd.json_normalize(data['items'])
        df = make_df(df)  # 제목과 요약에 있는 태그 제거 및 날짜 포맷 변환
        news_list = [(row['title'], row['link']) for _, row in df.iterrows()]
        return news_list
    else:
        return []

# 제미나이 API를 통해 답변 생성 및 키워드 추출 함수
def chatbot_response_via_gemini(user_input):
    try:
        # 제미나이 API를 통해 답변 생성
        response = model.generate_content(user_input)
        
        return response.text
    except Exception as e:
        return f"오류 발생: {str(e)}", None
    
# 제미나이 API를 통해 키워드를 추출하는 함수
def extract_keyword_via_gemini(com_input):
    try:
        # 제미나이 API를 통해 답변 생성 (키워드 추출용)
        response = model.generate_content(f'{com_input} 여기에서 중요한 키워드 하나만 단어 형태로 뽑아줘')
        
        # 중요한 키워드 추출 (간단한 예시로 첫 번째 단어 사용)
        keyword = response.text  # 예시로 첫 번째 단어 추출
        return keyword
    except Exception as e:
        return f"오류 발생: {str(e)}"

# 뉴스 링크에서 본문을 제미나이 API로 요약
def summarize_news_via_gemini(news_link):
    try:
        # 뉴스 링크에서 본문 추출 (간단히 뉴스 링크를 텍스트로 처리)
        response = requests.get(news_link)
        if response.status_code == 200:
            news_content = response.text  # 본문을 API로 받아올 수도 있음
            summary = model.generate_content(news_content).text
            return summary
        else:
            return "뉴스를 가져오지 못했습니다."
    except Exception as e:
        return f"오류 발생: {str(e)}"

# Gradio 인터페이스 정의
with gr.Blocks() as app:
    with gr.Row():
        # 왼쪽: 사용자 입력 및 제미나이 API 챗봇 답변
        with gr.Column(scale=1):  # 화면 반으로 나눔 (scale=1로 왼쪽 영역)
            user_input = gr.Textbox(label="질문 입력", placeholder="여기에 질문을 입력하세요...")
            submit_button = gr.Button("입력 완료")
            chatbot_output = gr.Textbox(label="챗봇 답변", placeholder="챗봇 답변이 여기에 표시됩니다.", interactive=False)
            keyword_output = gr.Textbox(label="추출된 키워드", placeholder="추출된 키워드가 여기에 표시됩니다.", interactive=False)
            
            # 입력 완료 버튼 클릭 시 제미나이 API를 통해 답변 생성 및 키워드 추출
            def on_submit(user_input):
                chatbot_reply = chatbot_response_via_gemini(user_input)
                keyword = extract_keyword_via_gemini(user_input)
                news_list = fetch_related_news(keyword)
                news_titles = [title for title, _ in news_list]  # 드롭다운에 보여줄 제목들
                return chatbot_reply, keyword, news_titles, dict(news_list)

            news_dropdown = gr.Dropdown(label="관련 뉴스", choices=[], interactive=True)
            news_summary = gr.Textbox(label="뉴스 요약", placeholder="선택된 뉴스의 요약이 여기에 표시됩니다.", interactive=False)
            
            # 질문 입력 후 관련 뉴스 가져오기
            submit_button.click(on_submit, inputs=user_input, outputs=[chatbot_output, keyword_output, news_dropdown])

        # 오른쪽: 추출된 키워드 및 관련 뉴스
        with gr.Column(scale=1):  # 오른쪽 영역 (scale=1)
            # 뉴스 선택 시 요약 표시
            def on_news_select(news_title, news_links):
                selected_news = news_links.get(news_title)
                if selected_news:
                    summary = summarize_news_via_gemini(selected_news)
                    return summary
                return "해당 뉴스를 찾을 수 없습니다."
            
            news_dropdown.change(on_news_select, inputs=[news_dropdown, news_dropdown], outputs=news_summary)

# Gradio 앱 실행
app.launch(inline=False, share=True)
```

```python
import gradio as gr
import os
import requests
import pandas as pd
from datetime import datetime

# 제미나이 모델 초기화 (모델 객체 생성)
model = genai.GenerativeModel("gemini-1.5-flash")

# 문자열에서 불필요한 태그 제거
def string_delete(text):
    return text.replace("<b>", "").replace("</b>", "").replace("&quot;", "")

# 날짜 문자열 파싱
def parse_date(date_str):
    return datetime.strptime(date_str, "%a, %d %b %Y %H:%M:%S %z")

# 데이터프레임에서 태그 제거 및 날짜 파싱
def make_df(df_name):
    for i in df_name:
        for idx in range(0, len(df_name[i])):
            df_name[i][idx] = string_delete(df_name[i][idx])
    df_name['pubDate'] = df_name['pubDate'].apply(parse_date)
    return df_name

# 네이버 뉴스 API에서 뉴스 목록 가져오기
def fetch_related_news(keyword):
    client_id = os.getenv('id')  # 환경 변수에서 네이버 클라이언트 ID 가져오기
    client_pw = os.getenv('pw')  # 환경 변수에서 네이버 클라이언트 Secret 가져오기
    
    url = "https://openapi.naver.com/v1/search/news.json"
    payload = {'query': keyword, 'display': 20, 'start': 1, 'sort': 'date'}  # 뉴스 검색 쿼리
    headers = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_pw}
    
    # API 요청
    r = requests.get(url, params=payload, headers=headers)
    if r.status_code == 200:
        data = r.json()
        df = pd.json_normalize(data['items'])
        df = make_df(df)  # 제목과 요약에 있는 태그 제거 및 날짜 포맷 변환
        news_list = [(row['title'], row['link']) for _, row in df.iterrows()]
        return news_list
    else:
        return []

# 제미나이 API를 통해 답변 생성 및 키워드 추출 함수
def chatbot_response_via_gemini(user_input):
    try:
        # 제미나이 API를 통해 답변 생성
        response = model.generate_content(user_input)
        return response.text
    except Exception as e:
        return f"오류 발생: {str(e)}"
    
# 제미나이 API를 통해 키워드를 추출하는 함수
def extract_keyword_via_gemini(com_input):
    try:
        # 제미나이 API를 통해 답변 생성 (키워드 추출용)
        response = model.generate_content(f'{com_input} 여기에서 중요한 키워드 하나만 단어 형태로 뽑아줘')
        keyword = response.text  # 제미나이 API로 추출된 키워드
        return keyword
    except Exception as e:
        return f"오류 발생: {str(e)}"

# 뉴스 링크에서 본문을 제미나이 API로 요약
def summarize_news_via_gemini(news_link):
    try:
        summary = model.generate_content(f'{news_link} 이 링크에 있는 뉴스 내용을 요약해줘')
        return summary.text
    except Exception as e:
        return f"오류 발생: {str(e)}"

# Gradio 인터페이스 정의
with gr.Blocks() as app:
    with gr.Row():
        # 왼쪽: 사용자 입력 및 제미나이 API 챗봇 답변
        with gr.Column(scale=1):  # 화면 반으로 나눔 (scale=1로 왼쪽 영역)
            user_input = gr.Textbox(label="질문 입력", placeholder="여기에 질문을 입력하세요...")
            submit_button = gr.Button("입력 완료")
            chatbot_output = gr.Textbox(label="챗봇 답변", placeholder="챗봇 답변이 여기에 표시됩니다.", interactive=False)
            keyword_output = gr.Textbox(label="추출된 키워드", placeholder="추출된 키워드가 여기에 표시됩니다.", interactive=False)
            
            # 입력 완료 버튼 클릭 시 제미나이 API를 통해 답변 생성 및 키워드 추출
            def on_submit(user_input):
                chatbot_reply = chatbot_response_via_gemini(user_input)
                keyword = extract_keyword_via_gemini(user_input)
                news_list = fetch_related_news(keyword)
                news_titles = [title for title, _ in news_list]  # 드롭다운에 보여줄 제목들
                news_links_dict = {title: link for title, link in news_list}  # 제목과 링크를 딕셔너리로 저장
                return chatbot_reply, keyword, gr.update(choices=news_titles), news_links_dict

            news_dropdown = gr.Dropdown(label="관련 뉴스", choices=[], interactive=True)
            news_summary = gr.Textbox(label="뉴스 요약", placeholder="선택된 뉴스의 요약이 여기에 표시됩니다.", interactive=False)
            
            # Gradio의 상태로 news_links_dict 저장
            news_links_state = gr.State()
            
            # 질문 입력 후 관련 뉴스 가져오기
            submit_button.click(on_submit, inputs=user_input, outputs=[chatbot_output, keyword_output, news_dropdown, news_links_state])

        # 오른쪽: 추출된 키워드 및 관련 뉴스
        with gr.Column(scale=1):  # 오른쪽 영역 (scale=1)
            # 뉴스 선택 시 요약 표시
            def on_news_select(news_title, news_links):
                selected_news = news_links.get(news_title)
                if selected_news:
                    summary = summarize_news_via_gemini(selected_news)
                    return summary
                return "해당 뉴스를 찾을 수 없습니다."
            
            news_dropdown.change(on_news_select, inputs=[news_dropdown, news_links_state], outputs=news_summary)

# Gradio 앱 실행
app.launch(inline=False, share=True)
```

```python
import gradio as gr
import os
import requests
import pandas as pd
from datetime import datetime

# 제미나이 모델 초기화 (모델 객체 생성)
model = genai.GenerativeModel("gemini-1.5-flash")

# 문자열에서 불필요한 태그 제거
def string_delete(text):
    return text.replace("<b>", "").replace("</b>", "").replace("&quot;", "")

# 날짜 문자열 파싱
def parse_date(date_str):
    return datetime.strptime(date_str, "%a, %d %b %Y %H:%M:%S %z")

# 데이터프레임에서 태그 제거 및 날짜 파싱
def make_df(df_name):
    for i in df_name:
        for idx in range(0, len(df_name[i])):
            df_name[i][idx] = string_delete(df_name[i][idx])
    df_name['pubDate'] = df_name['pubDate'].apply(parse_date)
    return df_name

# 네이버 뉴스 API에서 뉴스 목록 가져오기
def fetch_related_news(keyword):
    client_id = os.getenv('id')  # 환경 변수에서 네이버 클라이언트 ID 가져오기
    client_pw = os.getenv('pw')  # 환경 변수에서 네이버 클라이언트 Secret 가져오기
    
    url = "https://openapi.naver.com/v1/search/news.json"
    payload = {'query': keyword, 'display': 20, 'start': 1, 'sort': 'date'}  # 뉴스 검색 쿼리
    headers = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_pw}
    
    # API 요청
    r = requests.get(url, params=payload, headers=headers)
    if r.status_code == 200:
        data = r.json()
        df = pd.json_normalize(data['items'])
        df = make_df(df)  # 제목과 요약에 있는 태그 제거 및 날짜 포맷 변환
        news_list = [(row['title'], row['originallink']) for _, row in df.iterrows()]
        return news_list
    else:
        return []

# 제미나이 API를 통해 답변 생성 및 키워드 추출 함수
def chatbot_response_via_gemini(user_input):
    try:
        # 제미나이 API를 통해 답변 생성
        response = model.generate_content(user_input)
        return response.text
    except Exception as e:
        return f"오류 발생: {str(e)}"
    
# 제미나이 API를 통해 키워드를 추출하는 함수
def extract_keyword_via_gemini(com_input):
    try:
        # 제미나이 API를 통해 답변 생성 (키워드 추출용)
        response = model.generate_content(f'{com_input} 여기에서 중요한 키워드 하나만 단어 형태로 뽑아줘')
        keyword = response.text  # 제미나이 API로 추출된 키워드
        return keyword
    except Exception as e:
        return f"오류 발생: {str(e)}"

import requests
from bs4 import BeautifulSoup

# 뉴스 링크에서 본문을 추출하고 제미나이 API로 요약
def summarize_news_via_gemini(news_link):
    try:
        # 뉴스 링크에서 HTML 페이지 가져오기
        response = requests.get(news_link)
        if response.status_code != 200:
            return "뉴스를 가져오지 못했습니다."
        
        # BeautifulSoup을 사용하여 페이지 파싱
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # 네이버 뉴스 본문 추출 (네이버 뉴스는 'div' 태그의 class 'go_trans'에 본문이 있음)
        news_content = ''
        
        # 네이버 뉴스 본문이 들어 있는 태그 찾기
        content_div = soup.find('div', {'id': 'dic_area'})  # 뉴스 본문 영역
        
        if content_div:
            news_content = content_div.get_text(separator=' ', strip=True)  # 텍스트만 추출하여 공백으로 구분
        
        # 본문이 너무 길면 적절히 잘라서 제미나이 API로 보내기
        if len(news_content) > 5000:  # 제미나이 API에 전달 가능한 최대 텍스트 길이 (적절히 설정 필요)
            news_content = news_content[:5000]
        
        # 제미나이 API로 뉴스 본문 요약 요청
        summary = model.generate_content(f'{news_content} 이 뉴스 내용을 요약해줘')
        return summary.text
    
    except Exception as e:
        return f"오류 발생: {str(e)}"


# Gradio 인터페이스 정의
with gr.Blocks() as app:
    with gr.Row():
        # 왼쪽: 사용자 입력 및 제미나이 API 챗봇 답변
        with gr.Column(scale=1):  # 화면 반으로 나눔 (scale=1로 왼쪽 영역)
            user_input = gr.Textbox(label="질문 입력", placeholder="여기에 질문을 입력하세요...")
            submit_button = gr.Button("입력 완료")
            chatbot_output = gr.Textbox(label="챗봇 답변", placeholder="챗봇 답변이 여기에 표시됩니다.", interactive=False)
            keyword_output = gr.Textbox(label="추출된 키워드", placeholder="추출된 키워드가 여기에 표시됩니다.", interactive=False)
            
            # 입력 완료 버튼 클릭 시 제미나이 API를 통해 답변 생성 및 키워드 추출
            def on_submit(user_input):
                chatbot_reply = chatbot_response_via_gemini(user_input)
                keyword = extract_keyword_via_gemini(user_input)
                news_list = fetch_related_news(keyword)
                news_titles = [title for title, _ in news_list]  # 드롭다운에 보여줄 제목들
                news_links_dict = {title: link for title, link in news_list}  # 제목과 링크를 딕셔너리로 저장
                return chatbot_reply, keyword, gr.update(choices=news_titles), news_links_dict

            news_dropdown = gr.Dropdown(label="관련 뉴스", choices=[], interactive=True)
            news_summary = gr.Textbox(label="뉴스 요약", placeholder="선택된 뉴스의 요약이 여기에 표시됩니다.", interactive=False)
            
            # Gradio의 상태로 news_links_dict 저장
            news_links_state = gr.State()
            
            # 질문 입력 후 관련 뉴스 가져오기
            submit_button.click(on_submit, inputs=user_input, outputs=[chatbot_output, keyword_output, news_dropdown, news_links_state])

        # 오른쪽: 추출된 키워드 및 관련 뉴스
        with gr.Column(scale=1):  # 오른쪽 영역 (scale=1)
            # 뉴스 선택 시 요약 표시
            def on_news_select(news_title, news_links):
                selected_news = news_links.get(news_title)
                if selected_news:
                    summary = summarize_news_via_gemini(selected_news)
                    return summary
                return "해당 뉴스를 찾을 수 없습니다."
            
            news_dropdown.change(on_news_select, inputs=[news_dropdown, news_links_state], outputs=news_summary)

# Gradio 앱 실행
app.launch(inline=False, share=True)
```
