## 분석에 사용할 컬럼
- date : 분기 컬럼을 만들고 원-핫 인코딩으로 활용. 분석할 땐 date는 삭제
- 시군구 : 동제외, 자치구별_평균거래금액, 상세주소_평균거래금액에 활용. 분석할 땐 시군구는 삭제
- 단지명 : 분석할 땐 삭제
- 전용면적 : 전용면적은 사용할 예정이나 컬럼명에 특수문자가 들어가면 안되기에 컬럼명만 변경할 예정
- 거래금액 : 거래금액은 사용할 예정이나 컬럼명에 특수문자가 들어가면 안되기에 컬럼명만 변경할 예정
- 층 : 데이터 타입을 int로 바꾸고 사용
- 건축년도 : 아파트 나이 컬럼을 만들고 분석할 땐 삭제
- 학원수 : 그대로 사용

### 원핫인코딩 하고 스케일러 안하고 xgboost 돌려보기

```python
X2 = result.drop('거래금액', axis = 1)
y2 = result['거래금액']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor
```

```python
X2 = pd.get_dummies(X2, columns = ['분기', '동제외'])
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size = 0.3, random_state = 7)
```

```python
xgbr = XGBRegressor(max_depth = 5, n_estimators = 1000, n_jobs = 10, random_state = 7)
xgbr.fit(X2_train, y2_train)
pred2 = xgbr.predict(X2_test)
print("mean squared error : ", mean_squared_error(y2_test, pred2))
print("root mean squared error : ", np.sqrt(mean_squared_error(y2_test, pred2)))
print("r-squared : ", r2_score(y2_test, pred2))
```

```python
xgbr_result = pd.DataFrame(xgbr.feature_importances_, index = xgbr.feature_names_in_).sort_values(by = 0, ascending = False)
xgbr_result
```

```python
xgbr_result[xgbr_result[0]>0.01]
```

```python
plt.figure(figsize=(20, 10))
sns.regplot(x = y2_test, y = pred2, scatter = True)
plt.show()
```

```python
X2_importance = X2[['상세주소_평균거래금액', '분기_2009-Q1', '분기_2009-Q2', '동제외_서울특별시 서초구',
       '동제외_서울특별시 성동구', '전용면적', '기준금리', '분기_2024-Q3', '달러환율', '분기_2024-Q2',
       '동제외_서울특별시 노원구', '학원수', '동제외_경기도 성남시', '동제외_경기도 하남시', '동제외_부산광역시 강서구',
       '동제외_서울특별시 양천구', '분기_2008-Q4']]
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
X2_scaled = rs.fit_transform(X2_importance)
X2_scaled = pd.DataFrame(X2_scaled, columns = X2_importance.columns)
X2_scaled
```

```python
X2_scaled_train, X2_scaled_test, y2_scaled_train, y2_scaled_test = train_test_split(X2_scaled, y2, test_size = 0.3, random_state = 7)
```

```python
xgbr2 = XGBRegressor(max_depth = 5, n_estimators = 1000, n_jobs = 10, random_state = 7)
xgbr2.fit(X2_scaled_train, y2_scaled_train)
pred2_scaled = xgbr2.predict(X2_scaled_test)
print("mean squared error : ", mean_squared_error(y2_scaled_test, pred2_scaled))
print("root mean squared error : ", np.sqrt(mean_squared_error(y2_scaled_test, pred2_scaled)))
print("r-squared : ", r2_score(y2_scaled_test, pred2_scaled))
```
