```python
import urllib.request
from urllib.error import HTTPError
import time
import json

data_list = []
client_id = "Pv1RGKYi1IwvRpORRO_y" # 네이버 api에 접속 가능한 id
client_secret = "3DUDFWRLjC" # 네이버 api에 접속 가능한 pw
encText = urllib.parse.quote("파이썬")
a = True
page = 0
n = 1

try:
    while a:
        url = f"https://openapi.naver.com/v1/search/book.json?query={encText}&display=10&start={n}" # JSON 결과
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id", client_id)
        request.add_header("X-Naver-Client-Secret", client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
        if rescode == 200:
            response_body = response.read()
            data = json.loads(response_body.decode('utf-8'))
            data_list.append(data)
        else:
            print("Error Code:", rescode)
        page += 1
        n += 10
        print('page :', page, end='\r')
        time.sleep(0.5)
        if page >= data['total'] // 10 + 1:
            a = False

except HTTPError as e:
    print(f"HTTP error occurred: {e.code} - {e.reason}")
except Exception as e:
    print(f"An error occurred: {e}")

print("Data collection completed.")
print(data_list)
```
