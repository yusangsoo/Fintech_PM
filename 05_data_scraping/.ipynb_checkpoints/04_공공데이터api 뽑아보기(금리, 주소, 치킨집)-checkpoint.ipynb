### 치킨집만 뽑아오기

```python
# 1페이지만
url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
servicekey = os.getenv('service_key')
payload = {
    'servicekey' : servicekey,
    'divId' : 'indsSclsCd',
    'key' : 'I21006',
    'pageNo' : 1,
    'numOfRows' : 10,
    'type' : 'json'
    
}

r = requests.get(url, params = payload)
if r.status_code == 200 :
    data = r.json()
    print(data)
else :
    print("Error Code:", r.status_code)
```

```python
print(data['body'].keys())
print(data['body']['totalCount'])
```

```python
# while 문으로 전체 페이지 수집
a = True
page = 1
data_list = []
while a :
    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    servicekey = os.getenv('service_key')
    payload = {
        'servicekey' : servicekey,
        'divId' : 'indsSclsCd',
        'key' : 'I21006',
        'pageNo' : page,
        'numOfRows' : 1000,
        'type' : 'json'
    }

    r = requests.get(url, params = payload)
    
    if page <= data['body']['totalCount'] // 1000 + 1 : 
        page += 1
        print('page : ', page, end = '\r')
    else :
        a = False
        
    if r.status_code == 200 :
        data = r.json()
        data_list.append(data)
    else :
        print("Error Code:", r.status_code)
print(data_list[0])
```

```python
# 데이터프레임으로 만들기
result = pd.DataFrame()
for i in data_list :
    temp = pd.json_normalize(i['body']['items'])
    result = pd.concat([result,temp])
result
```
