## api 명세서를 참고해서 key 리스트를 만들고 딕셔너리 함수 활용해 만들기

강사님과 함께 한 코드

```python
# 1페이지만
url = "http://apis.data.go.kr/1160100/service/GetSmallLoanFinanceInstituteInfoService/getOrdinaryFinanceInfo"
servicekey = os.getenv('service_key')
payload = {
    'serviceKey' : servicekey,
    'pageNo' : 1,
    'numOfRows' : 10,
    
}

r = requests.get(url, params = payload)
data = r.text
soup = BeautifulSoup(data, 'xml')
```

```python
# 여기까진 나도 함.
# 이후에 키와 해당하는 밸류를 한 번에 만들기에 실패함.
# 할려면 할 순 있었음. 결과물을 하나의 리스트로 만들고 10개씩 끊어서 각 리스트로 만들어서 직접 넣으면 됨.
# 그러나 더 효율적으로 하고 싶었음.
for key in tag_name :
    for item in soup.find_all(key) :
        print(item.string)
```

```python
# 내가 실패한 딕셔너리로 만들기
# setdefault를 생각하지 못함.
result = {}
for key in tag_name :
    for item in soup.find_all(key) :
        result.setdefault(key,[]).append(item.string)
result
```

```python
pd.DataFrame(result)
```

## api응답에서 xml 태그를 추출해 자동으로 데이터 정리하기

```python
# 전체 페이지
page = 1
a = True
url = "http://apis.data.go.kr/1160100/service/GetSmallLoanFinanceInstituteInfoService/getOrdinaryFinanceInfo"
servicekey = os.getenv('service_key')
total_page = int(soup.find('totalCount').string) // int(soup.find('numOfRows').string) + 1
result = {}
keys = {}
while a :
    payload = {
        'serviceKey' : servicekey,
        'pageNo' : page,
        'numOfRows' : 1000,
    }

    # beatuifulsoup을 활용해 xml로 파싱하기 = xml 파일을 읽을 수 있도록 가져오는 행위
    r = requests.get(url, params = payload)
    data = r.text
    soup = BeautifulSoup(data, 'xml')
    
    
    # soup.find.all()의 결과는 header부터 가져오기 때문에 우리가 필요한 태그만 가져오기 위함
    # 태그 이름을 딕셔너리의 키값으로 지정
    for tags in soup.find_all()[10::] :
        if tags.name == 'item' :
            continue
    #     print(tags.name)
        else :
            keys.setdefault(f'{tags.name}',"")
    
    # 전체 페이지를 가져오기 위해서 현재 페이지가 전체 페이지보다 낮을 경우만 진행하도록 함
    if page < total_page :
        print('page : ', page)
        page += 1
    else :
        a = False


    # 태그 이름을 딕셔너리의 키 값으로, 태그 값의 문자를 딕셔너리의 밸류로 지정
    for key in keys :
        for item in soup.find_all(key) :
            result.setdefault(key,[]).append(item.string)

final_df = pd.DataFrame(result)
# for df in result :
#     final_df = pd.concat([final_df, df])
final_df
```
