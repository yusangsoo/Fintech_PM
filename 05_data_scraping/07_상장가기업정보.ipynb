### 대한민국 대표 기업공시채널에서 상장법인목록 가져오기

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://kind.krx.co.kr/corpgeneral/corpList.do"
payload = {'method' : 'loadInitPage'}
r = requests.get(url, params = payload)
print(r.text)
```

```python
data = r.text
soup = BeautifulSoup(data, 'xml')
soup
```
